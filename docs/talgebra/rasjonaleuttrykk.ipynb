# Rasjonale uttrykk

## Grunnleggende brøkregning
Når vi arbeider med brøker blir vi ofte nødt til å utvide eller forkorte en brøk. Det er viktig å huske på at det ikke endrer verdien til brøken. For å forkorte en brøk lønner det seg å starte med å faktorisere teller (tallet over brøkstreken) og nevner (tallet under brøkstreken) for så å forkorte vekk felles faktorer.

```{admonition} Utvide og forkorte en brøk
:class: eksempel, dropdown

**Utviding av brøk**

$\frac{1}{2}=\frac{1\cdot 5}{2\cdot 5}=\frac{5}{10}$

**Forkorting av brøk**

$\frac{84}{420}=\frac{2\cdot 2\cdot 3\cdot 7}{2\cdot 2\cdot 3\cdot 5\cdot 7}=\frac{1}{5}$
```

Når vi skal addere eller subtrahere med brøk må vi først utvide alle brøkene til fellesnevner. Fellesnevneren er et tall som alle nevnerene går opp i (fellesnevneren kan divideres på alle de opprinnelige nevnerne). Vi vil alltid finne en fellesnevner ved å multiplisere alle nevnerene, men som oftest vil vi kunne finne enklere fellesnevnere enn det. Vi kommer tilbake til teknikker for å finne fellesnevner når vi ser på rasjonale uttrykk lenger ned på sida.

```{admonition} Addisjon og subtraksjon av brøk
:class: eksempel, dropdown

$$
\begin{align}
\frac{5}{12}+\frac{1}{4}-\frac{1}{3} &=\frac{5}{12}+\frac{3}{12}-\frac{4}{12} \\
&= \frac{5+3-4}{12} \\ 
&=\frac{4}{12} \\
&= \frac{1}{3}
\end{align}
$$
```

Multiplikasjon med brøk utføres ved å multiplisere teller med teller og nevner med nevner. Et viktig spesialtilfelle er multiplikasjon av brøk med heltall. Det kan vi utføre ved å gjøre om heltallet til en brøk med $1$ i nevneren. Da ser vi at det effektivt sett blir det samme som å multiplisere heltallet inn i telleren.

```{admonition} Multiplikasjon med brøk
:class: eksempel, dropdown

**Multiplikasjon av to brøker**

$
\frac{1}{4}\cdot\frac{2}{3} =\frac{1\cdot 2}{4\cdot 3}=\frac{2}{12}=\frac{1}{6}
$

**Multiplikasjon av brøk med heltall**

$
5\cdot\frac{2}{3}=\frac{5}{1}\cdot\frac{2}{3}=\frac{5\cdot 2}{1\cdot 3}=\frac{10}{3}
$

```

Divisjon med brøk utføres ved å multiplisere med den omvendte brøken. At det må være slik ser vi ved å skrive det som en brudden brøk (brøk med brøk) og utvide.

```{admonition} Divisjon med brøk
:class: eksempel, dropdown

$$
\frac{1}{4} : \frac{2}{3}=\frac{1}{4}\cdot \frac{3}{2}=\frac{1\cdot 3}{4\cdot 2}= \frac{3}{8}
$$

**Forklaring med brudden brøk**

$$
\frac{\frac{1}{4}}{\frac{2}{3}}=\frac{\frac{1}{4}\cdot 4\cdot 3}{\frac{2}{3}\cdot 4\cdot 3}=\frac{1\cdot 3}{4\cdot 2}=\frac{1}{4}\cdot \frac{3}{2}
$$
```

Det er også verdt å minne om at en andel skrevet som brøk kan gjøres om til henholdsvis desimaltall og prosentandel.

```{admonition} Brøk, desimaltall og prosent
:class: eksempel, dropdown

$\frac{1}{5}=0,20=20 \%$
```

## Rasjonale uttrykk
Et rasjonalt uttrykk er brøker med algebraiske uttrykk i teller og eller nevner. Vi kommer til å møte på rasjonale uttrykk i mange situasjoner, så vi tar litt tid til å jobbe spesielt med dem. Framgangsmåtene vi brukte over for brøker med rene tall gjelder også for rasjonale uttrykk. Vi skal se en del eksempler på det, men først skal vi konsentrere oss om en metode for å finne fellesnevner.

### Finne fellesnevner

Vi kan alltid finne en fellesnevner ved å multiplisere sammen alle nevnerne i brøkene vi skal arbeide med. Dersom nevnerne kun består av tall, vil dette ofte fungere greit. Når vi har algebraiske uttrykk i nevneren, vil denne metoden være unødvendig komplisert. Vi trenger derfor en god metode for å finne den enkleste fellesnenveren.

```{admonition} Enkleste fellesnevner
Metode 1:
 * Faktoriser alle nevnerne
 * Vi starter med å la fellesnevneren være den første nevneren
 * Så går vi gjennom de andre nevnerene og legger til faktorer i fellesnevneren som den ikke allerede har

Metode 2:
 * Faktoriser alle nevnerne
 * Skriv opp alle unike faktorer som forekommer i nevnerne
 * For faktorer som det er flere *kopier* av i en nevner må vi ha med så mange som det er i den nevneren som har flest av den
```

Metodene er kanskje enklest å forstå gjennom eksempel.

```{admonition} Eksempel 1: Fellesnevner
:class: eksempel, dropdown
Finn fellesnevneren til $\frac{x}{4x^2}$ og $\frac{1}{8x}$

For begge metodene må vi faktorisere nevnerne:

$4x^2=2\cdot 2\cdot x\cdot x$
$8x = 2\cdot 2\cdot 2\cdot x$

Metode 1:
Vi starter med å la fellesnevneren være den første nevneren $FN = 2\cdot 2\cdot x\cdot x$.
Så ser vi på den andre nevneren og ser at vi trenger en ekstra faktor 2 slik at $FN = 2\cdot 2\cdot 2\cdot x\cdot x = 8x^2$

Metode 2:
Vi ser at fellesnevneren må inneholde faktorene $2$ og $x$. Den andre brøken har flest faktorer av $2$, mens den første har flest faktorer av $x$. Vi må derfor har med tre faktorer av $2$ og to faktorer av $x$, slik at $FN = 2\cdot 2\cdot 2\cdot x\cdot x = 8x^2$.
```

```{admonition} Eksempel 2: Fellesnevner
:class: eksempel, dropdown
Finn fellesnevneren til $\frac{x}{2x^2+2x}$, $\frac{1}{x^2+2x+1}$ og $\frac{3x}{x^2-1}$.

Vi starter med å faktorisere nevnerne:

$2x^2+2x=2x(x+1)$
$x^2+2x+1=(x+1)^2$
$x^2-1=(x+1)(x-1)$

Uansett valg av metode, så ser vi at vi må ha en faktor $2$, en faktor $x$, to faktorer $(x+1)$ og en faktor $(x-1)$.

Fellesnevneren blir: $2x(x-1)(x+1)^2$
```

Når vi har kontroll på faktorisering og å finne fellesnevner er resten mye enklere.

### Eksempler på regning med rasjonale uttrykk
Generelt for alle eksemplene under er at vi starter med å faktorisere nevner i alle brøkene. Ofte vil det også være en fordel å faktorisere tellerne allerede fra start. Vi starter med multiplikasjon og divisjon som er de to enkleste operasjonene med brøk.

```{admonition} Eksempel 1: Multiplikasjon
:class: eksempel, dropdown

$\frac{8}{x^2}\cdot\frac{5x}{6}=\frac{2^3\cdot 5\cdot x}{x^2\cdot 2\cdot 3}=\frac{2^2\cdot 5}{x\cdot 3}=\frac{20}{3x}$
```

```{admonition} Eksempel 2: Multiplikasjon
:class: eksempel, dropdown

$\frac{36-a^2}{12a^2}\cdot\frac{4a}{6-a}=\frac{(6-a)(6+a)\cdot 4 \cdot a}{3\cdot 4\cdot a^2\cdot (6-a)}=\frac{6+a}{3a}$
```

```{admonition} Eksempel 1: Divisjon
:class: eksempel, dropdown

$\frac{12}{x-2}:\frac{6}{4x-8}=\frac{3\cdot 4}{x-2}\cdot \frac{4(x-2)}{2\cdot 3}=\frac{3\cdot 4\cdot 4\cdot (x-2)}{(x-2)\cdot 2\cdot 3}=\frac{8}{1}=8$
```

```{admonition} Eksempel 2: Divisjon
:class: eksempel, dropdown

$\frac{25-a^2}{12a^2}:\frac{5+a}{4a}=\frac{(5-a)(5+a)}{3\cdot 4\cdot a^2}\cdot \frac{4\cdot a}{5+a}=\frac{(5-a)(5+a)\cdot 4a}{3\cdot 4\cdot a^2\cdot (5+a)}=\frac{5-a}{3a}$
```

Til slutt tar vi noen eksempler med addisjon og subtraksjon.

```{admonition} Eksempel 1: Addisjon og subtraksjon
:class: eksempel, dropdown

$\frac{12}{x}+\frac{3x}{3x^2}$

Ser først at fellesnevner er $3x^2$. Vi må da utvide den første brøken ved å multiplisere med $3x$ over og under brøkstreken.

$\frac{3x\cdot 12}{3x\cdot x}+\frac{3x}{3x^2}=\frac{36x}{3x^2}+\frac{3x}{3x^2}=\frac{36x+3x}{3x^2}=\frac{39x}{3x^2}$

Til slutt må vi undersøke om vi kan forkorte uttrykket. Her ser vi at i kan forkorte med $3x$. Om det er vanskelig å se, så hjelper det å faktorisere teller og nevner for så å forkorte felles faktorer.

$\frac{39x}{3x^2}=\frac{3\cdot 13 \cdot x}{3\cdot x\cdot x} = \frac{13}{x}$
```

```{admonition} Eksempel 2: Addisjon og subtraksjon
:class: eksempel, dropdown

$\frac{5y+25}{y^2-25}+\frac{y}{y-5}$

Ser først at $y^2-25=(y+5)(y-5)$. Dette blir fellesnevner og vi må utvide den andre brøken med $(y+5)$

$\frac{5y+25}{(y+5)(y-5)}+\frac{y\cdot(y+5)}{(y-5)(y+5)} = \frac{5y+25+y(y+5)}{(y+5)(y-5)}=\frac{y^2+10y+25}{(y+5)(y-5)}$

Til slutt må vi undersøke om vi kan forkorte uttrykket. Vi faktoriserer teller ved hjelp av første kvadratsetning.

$\frac{(y+5)^2}{(y+5)(y-5)}=\frac{y+5}{y-5}$
```
