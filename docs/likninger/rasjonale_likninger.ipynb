# Rasjonale likninger
Likningene vi har arbeidet med fram til nå har bestått av polynomuttrykk. Det vil si at den ukjente størrelsen kun har forekommet med positiv eksponent ($x$, $x^2$, $x^3$ og så videre). I en del tilfeller vil vi imidlertid møte likninger der den ukjente er i nevneren i uttrykket. Slike likninger krever at vi er litt ekstra forsiktig når vi løser dem. Vi kaller de for rasjonale likninger.

```{admonition} Rasjonale likninger
Likninger som inneholder rasjonale uttrykk kalles rasjonale likninger. Et rasjonalt uttrykk er et uttrykk der variabelen ($x$) befinner seg i nevneren i uttrykket.
```

Vi skal på den sida her lære hvordan vi kan løse slike likninger, men først vil vi repetere og utdype litt generelt om det vi kan om likningsløsning fra før.Grunnprinsippet for å løse likninger er at vi må gjøre samme regneoperasjon på begge sider av likhetstegnet fram til vi har $x$ (den ukjente størrelesen) alene på den ene siden av likhetstegnet. Det er det vi har brukt for å løse lineære likninger og det er det som er brukt for å utlede abc-formelen for andregradslikninger. Grunnen til at dette prinsippet holder er at så lenge vi utfører samme regneoperasjon på begge sider av et likhetstegn, så vil den nye og den opprinnelige likningen være ekvivalente. Vi kan for eksempel skrive:

$$\begin{align}
x+5&=6 \\
&\Updownarrow \\
x & = 1
\end{align}
$$

De to likningene har akkurat samme løsning og er derfor ekvivalente.

Det finnes imidlertid noen operasjoner vi kan gjøre som ikke bevarer ekvivalens mellom likningene. De viktigste av disse operasjonene er å multiplisere eller dividere med 0. Dersom vi multipliserer begge sider av ei likning med 0, så vil begge sider bli 0 uavhengig av hva de opprinnelig var. Det vil sjelden/aldri være et problem at vi multipliserer/dividerer med 0 direkte, men dersom vi multipliserer/dividerer med et uttrykk som inneholder $x$ vil det kunne skje indirekte.

```{admonition} Multiplisere/dividere med x
Dersom vi multipliserer med uttrykket $x-a$ ($a\in\mathbf{R}$) på begge sider av ei likning, så vil vi kunne innføre ei ny løsning $x=a$ i likningen.

Dersom vi dividerer med uttrykket $x-a$ ($a\in\mathbf{R}$) på begge sider av ei likning, så vil vi kunne fjerne ei løsning $x=a$ fra likningen.

I begge tilfeller vil regneoperasjonen gjøre at vi ikke har ekvivalens mellom den nye og den opprinnelige likningen. I stedet for vil vi ha en implikasjon der den opprinnelige likningen impliserer den nye.
```

Vi starter med et eksempel der vi multipliserer med $x$ uten å trenge

```{admontition} Eksempel: Multiplisere med x
:class: eksempel

Vi skal løse likningen $x + 5 = 6$. Det er ei enkel likning som vi ser har løsningen $x=1$. Men hva skjer om vi multipliserer begge sider av likningen med $x-3$? Da ender vi opp med likningen $(x-3)(x+5)=(x-3)\cdot 6$. Etter å ha multiplisert ut parentesene og flyttet alt til venstre side ender vi opp med $x^2-4x+3=0$. Dette er ei andregradslikning med løsningene $x=3$ og $x=0$. Vi ser at vi ved å multiplisere med $x-3$ har innført ei ekstra løsning $x=3$ i likningen. Det er derfor ikke ekvivalens mellom $x+5=6$ og $(x-3)(x+5)=(x-3)\cdot 6$. Den første medfører (impliserer) den andre, men de gir ikke eksakt samme løsning.
```

```{admonition} Eksempel: Dividere med x
:class: eksempel

Vi skal løse likningen $(x-2)(x+4)=(x-2)$. 

Likninga kan løses ved å multiplisere ut parentesene og flytte alt til venstre side. Da ender vi opp med likningen $x^2+x-6=0$ som har løsningene $x=-3$ og $x=2$.

Alternativt så kan vi tenke oss at vi dividerer begge sider med $x-2$ og ender opp pmed likningen $x+4=-3$. Den har løsningen $x=2$. Her ser vi at vi har mistet løsningen $x=2$ ved å dividere med $x+2$. Likningen $x+4=-3$ er derfor ikke ekvivalent med den opprinnelige likningen.
```

Når vi skal løse rasjonale likninger kan vi bli nødt til å multiplisere begge sidene av likningen med et uttrykk som inneholder $x$. I slike tilfeller må vi alltid kontrollere svaret for å se om vi har innført en ny løsning som ikke passer med den opprinnelige likningen. Vi ser dette best med noen eksempler:

```{admonition} Eksempel 1: Rasjonal likning
:class: eksempel
Vi skal løse likningen $\displaystyle{\frac{x-3}{x-2}-4=-\frac{1}{x-2}}$. For å løse likninger med brøk multipliserer vi gjerne med fellesnevner til begge sider for å bli kvitt brøkene. Her er fellesnevner $x-2$. Ved å multiplisere med dette vil vi ikke ha ekvivalens mellom likningene og må derfor gå tilbake å sjekke svaret til slutt.

$$\begin{align}
\frac{x-3}{x-2}-4&=-\frac{1}{x-2} \\
\frac{(x-2)(x-3)}{x-2}-4(x-2) &= -\frac{(x-2){x-2}} \\
(x-3)-4x+8 &= -1 \\
-3x &= -6 \\
x&=2
\end{align}
$$

Vi ender opp med løsninga $x=2$. Går vi tilbake til den opprinnelige likningen ser vi imidlertid at løsninga ikke er gyldig siden den medfører at det blir null i nevnerne. Grunnen til at vi allikevel får løsningen er at vi ved å multiplisere med $x-2$ i dette tilfellet har innført $x=2$ som løsning i likningen.

Løsninga av likninga er $L= \varnothing$.

Så når vi multipliserer med et ukjent uttrykk må vi sjekke svaret. Ei lita endring på likninga til $\frac{x-3}{x-2}-3=\frac{1}{x-2}$ løst på akkurat samme måte gir ei godkjent løsning $x=1$.
```

```{admonition} Eksempel 2: Rasjonal likning
:class: eksamen
Vi skal løse likningen $\displaystyle{\frac{1}{x-4}-\frac{x-3}{3x}=\frac{4}{x^2-4x}}$.

Det første vi må gjøre er å finne fellesnevner. Om man ikke ser det direkte, så kan vi faktorisere alle nevnere og bruke en av metodene fra siden om rasjonale uttrykk. Vi finner fellesnever lik $3x(x-4)$ og multipliserer alle ledd på begge sider av likningen med dette. Da forutsetter vi at $x\neq 0$ og $x\neq 4$ siden de løsningene medfører at vi multipliserer med null.

$$
\begin{align}
\frac{1}{x-4}-\frac{x-3}{3x}&=\frac{4}{x^2-4x} \\
\frac{3x(x-4)}{x-4}-\frac{3x(x-4)(x-3)}{3x}&=\frac{3x(x-4)\cdot 4}{x^2-4x} \\
3x-(x-4)(x-3) &= 12 \\
-x^2+10x-24&=0
\end{align}
$$

Løser andregradslikninga med abc-formelen og får $x=4 \vee x=6$. Siden vi har multiplisert med et uttrykk som inneholder $x$ må vi gå tilbake å kontrollere svaret. Vi ser at $x=4$ ikke er ei mulig løsning siden vi da dividerer med null i opprinnelig likning.

Løsninga av likninga er $L=\{6\}$.
```
