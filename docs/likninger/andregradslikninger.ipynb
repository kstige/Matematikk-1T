# Andregradslikninger

Fram til nå har vi arbeidet med lineære likninger. Det er likninger der den ukjente kun er i første potens. Skal vi bruke likninger til å løse problemstillinger innenfor andre fagfelt, så vil vi imidlertid ende opp med likninger der den ukjente er opphøyd i andre. Eksempler på slike likninger er

 1) $x^2=36$
 2) $2x^2-8=0$
 3) $x^2-12x=0$
 4) $x^2+6x+9=4$
 5) $x^2-12x+20=0$


```{admonition} Andregradslikninger
I en andregradslikning vil den ukjente størrelsen være i andre potens. Alle andregradslikninger kan skrives på formen

$$ax^2+bx+c=0$$

der $a, b, c \in \mathbf{R}$ og $a\neq 0$.
```

Dersom vi flytter alle ledd over på venstre side, så vil alle eksemplene over være på denne generelle formen. I de tre første er enten $b$ eller $c$ lik null. Vi starter med å se på disse tilfellene.

## $ax^2+c=0$ 

Vi starter med andregradslikninger der $b=0$. Da er likningen på formen $ax^2+c=0$. Slike likninger løser vi ved enklest ved å flytte $c$ over til høyre side, dividere på $a$ og så ta kvadratrot. I det første eksempelet viser vi hvordan vi løser den generelt før vi løser 1) og 2) i lista øverst på sida.

```{admonition} Eksempel: $ax^2+c=0$
:class: eksempel
$
\begin{align}
ax^2+c &= 0 \\
ax^2 &= -c \\
x^2 &= -\frac{c}{a} \\
\sqrt{x^2} &= \sqrt{-\frac{c}{a}} \\
|x| &= \sqrt{-\frac{c}{a}} \\
x &= \pm \sqrt{-\frac{c}{a}}
\end{align}
$

Her har vi brukt at $\sqrt{x^2}=|x|$ siden vi har definert kvadratroten til å være et positivt tall. 

Vi husker at vi kun kan ta kvadratroten av positive tall. Dersom vi skal kunne regne ut kvadratroten på høyre side må derfor enten $a$ eller $c$ være negativ (men ikke begge). Om $a$ og $c$ har likt fortegn vil vi få et negativt tall inne i kvadratoten og den er da ikke definert. I slike tilfeller har likningen ingen løsning. Videre vil vi ha en løsning dersom $c=0$ og to løsninger (gitt ved $\pm$ foran kvadratroten) dersom $a$ og $c$ har forskjellige fortegn.
```

Det vi ser over gjelder generelt for alle andregradslikninger.

```{admonition} Antall løsninger
En andregradslikning kan ha

 * ingen løsning
 * en løsning
 * to løsninger
```

Vi tar nå to talleksemepel

```{admonition} $x^2=36$
:class: eksempel

I denne likningen er vi klar til å ta kvadratrot med en gang

$
\begin{align}
x^2 &= 36 \\
\sqrt{x^2} &= \sqrt{36} \\
|x| &= 6 \\
x &= \pm 6
\end{align}
$

Det er også vanlig å skrive opp løsningen som $L=\{-6, 6\}$ som betyr at løsningsmengden til likningen er $\{-6, 6\}$

Vanligvis vil vi hoppe over linje 3 i mellomregningen her.
```

```{admonition} $2x^2-8=0$
:class: eksempel

$
\begin{align}
2x^2 -8 &= 0 \\
2x^2 &= 8 \\
x^2 &= 4 \\
\sqrt{x^2} &= \sqrt{4} \\
|x| &= 2 \\
x &= \pm 2
\end{align}
$

Løsningsmengden er $L=\{-2,2\}$.
```
