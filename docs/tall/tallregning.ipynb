# Tallregning

Vi skal her raskt repetere litt om grunnleggende tallregning.

## Regnerekkefølge

De fire grunnleggende regneoperasjonene er addisjon, subtraksjon, multiplikasjon og divisjon. I tillegg kommer vi til å ha bruk for potenser og røtter (kvadratrøtter).

Når vi skal utføre et sammensatt regnestykke, så benytter vi oss av denne regnerekkefølgen

```{admonition} Regnerekkefølgen

 1) Parenteser
 2) Potenser og røtter
 3) Multiplikasjon og divisjon
 4) Addisjon og subtraksjon

```

Vi minner også om noen viktige regneregler for reelle tall.

```{admonition} Regneregler for reelle tall
:class: dropdown

Vi lar $a, b \in \mathbf{R}$.

$a+b=b+a$

$a\cdot b = b\cdot a$

$a\cdot(b \pm c)= a\cdot b \pm a\cdot c$

$(-a)\cdot b = -a\cdot b$

$(-a) \cdot (-b) = a\cdot b$

De to første bruker vi stort sett uten å tenke over dem. Det er den *kommutative* egenskapen til addisjon og multiplikasjon. Vi kan addere og multiplisere i den rekkefølgen vi ønsker.

Nummer tre er den såkalte *assosiative* egenskapen til multiplikasjon med en parentes. $\pm$ inne i parentesen betyr at egenskapen gjelder uansett om det står $+$ eller $-$.

De to siste reglene er det vi husker som at multiplikasjon (eller divisjon) med samme fortegn gir positivt svar, mens multiplikasjon/divisjon med ulike fortegn gir negativt svar.
```


## Potenser og standardform
En potens er en gjentatt multiplikasjon.

```{admonition} Definisjon av potens
Dersom $a\in\mathbf{R}$ og $n\in\mathbf{N}$, så betyr $a^n$ at vi skal multiplisere $a$ med seg selv $n$ ganger. $a$ kalles grunntallet, mens $n$ er eksponenten.

$$a^n=\underbrace{a\cdot a \cdots a}_{n \text{ ganger}}$$
```

All regning med potenser kan utføres fra definisjonen, men det blir mye mer effektivt dersom vi tar i bruk regnereglene for potenser. Vi beviser ikke regnereglene her, men de fleste faller ganske logisk ut fra definisjonen over.

```{admonition} Regneregler for potenser

 1) $a^n \cdot a^m = a^{n+m}$
 2) $\frac{a^n}{a^m} = a^{n-m}$
 3) $\left(a\cdot b\right)^n = a^n \cdot a^m$
 4) $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$
 5) $\left(a^n\right)^m = a^{n\cdot m}$
 6) $a^0 = 1$
 7) $a^{-n} = \frac{1}{a^n}$

```

Når vi arbeider med tall som er veldig store, eller med tall som er veldig små kan en spesiell skrivemåte med potenser hjelpe oss.

```{admonition} Standardform
Standardform for et tall er

$a\cdot 10^n$, der $n\in \mathbf{Z}$ og $|a|\in[1,10\rangle$.

Vi skriver altså tallet med et siffer (1 eller større) før komma og multipliserer med en tierpotens for å beksrive størrelsen på tallet.
```

Standardform forklares enklest med noen eksempler.

```{admonition} Eksempel: Skrive tall på standardform
:class: eksempel, dropdown

For å skrive et tall på standardform teller vi antall plasser vi må flytte komma for å ha kun et siffer større eller lik 1 før komma. Eksponenten til tierpotensen blir antall plasser vi har flyttet komma mot venstre.

$53600000=5,36\cdot 10^7$

Her har vi flyttet komma 7 plasser mot venstre.

$0,0000000345 = 3,45\cdot 10^{-8}$

Her har vi flyttet komma 8 plasser mot høyre og får en negativ eksponent. Det er fordi $0,0000000345=\frac{3,45}{10^8}$.
```

```{admonition} Eksempel: Regne med potenser og standardform
:class: eksempel, dropdown

$\frac{160000\cdot 0,02}{0,008}=\frac{16\cdot 10^4\cdot 2\cdot 10^{-2}}{8\cdot 10^{-3}}=\frac{16\cdot 2}{8}\cdot 10^{4+(-2)-(-3)}=4\cdot 10^5$

Her ar vi ikke brukt skikkelig standardform før i svaret, men vi ser hvordan regnestykket forenkles ved å bruke tierpotenser og potensreglene.
```

## Kvadratrøtter
De fleste vet nok at $\sqrt{16} = 4$, men det kan være greit å definere begrepet kvadratrot skikkelig siden vi kommer til å benytte det inn mot blant annet andregradslikninger senere.

```{admonition} Definisjon av kvadratrot
Vi lar $a$ være et positivt reelt tall. Kvadratroten av $a$ er det positive tallet som multiplisert med seg selv gir $a$.

$$\left(\sqrt{a}\right)^2=a$$
```

I definisjonen over er det to ting som er viktig å legge merke til.
 1) Vi kan kun ta kvadratroten av positive reelle tall.
 2) Kvadratroten er et positivt tall. Selv om både $(-4)^2=16$ og $4^2=16$, så er $\sqrt{16}= 4$.

Generelt kan vi skrive at $\sqrt{a^2}=|a|$. Der $|a|$ er absoluttverdien til tallet $a$. Med absoluttverdien mener vi avstanden tallet har fra $0$ på tallinja. Vi kan derfor si at $|-4|=|4|=4$ siden begge tall har avstanden $4$ fra $0$.

Om kvadratrøtter kan det videre være nyttig å vite følgende.

```{admonition} Viktig om kvadratrøtter
$$\sqrt{a+b}\neq \sqrt{a}+\sqrt{b}$$

$$\sqrt{a\cdot b} = \sqrt{a}\cdot \sqrt{b}$$
```

## Primtall og faktorisering

Et primtall er et tall som kun kan divideres på seg selv og $1$. De første primtallene er $2, 3, 5, 7, 11 og 13$. Det kan bevises at det finnes uendelig antall primtall. 

Alle tall som ikke er primtall (bortsett fra $1$) kaller vi for sammensatte tall. Sammensatte tall kan vi skrive som et produkt av andre tall. Det kalles å faktorisere tallet. Et fundamentalt teorem i matematikken sier at alle tall kan skrives som et produkt av primtall på en unik måte. En slik faktorisering kalles *primtallsfaktorisering*. Vi kan for eksempel primtallsfaktorisere $84=2\cdot 2\cdot 3\cdot 7$.

Faktorisering blir et viktig redskap når vi skal arbeide med likninger. Her nøyer vi oss med å se et eksempel på hvordan vi kan bruke faktorisering til å forenkle uttrykk med kvadratrøtter.

```{admonition} Eksempel: Forenkling av kvadratrøtter
:class: eksempel
$$
\begin{align}
\sqrt{250}-\sqrt{40} &=\sqrt{25\cdot 10}-\sqrt{4\cdot 10}\\
&= \sqrt{25}\cdot\sqrt{10}-\sqrt{4}\cdot\sqrt{10} \\
&=5\cdot\sqrt{10}-2\cdot\sqrt{10} \\
&=3\sqrt{10}
\end{align}
$$
```

## Brøkregning
Når vi arbeider med brøker blir vi ofte nødt til å utvide eller forkorte en brøk. Det er viktig å huske på at det ikke endrer verdien til brøken. For å forkorte en brøk lønner det seg å starte med å faktorisere teller og nevner for så å forkorte vekk felles faktorer.

```{admonition} Utvide og forkorte en brøk
:class: eksempel, dropdown

**Utviding av brøk**

$\frac{1}{2}=\frac{1\cdot 5}{2\cdot 5}=\frac{5}{10}$

**Forkorting av brøk**

$\frac{84}{420}=\frac{2\cdot 2\cdot 3\cdot 7}{2\cdot 2\cdot 3\cdot 5\cdot 7}=\frac{1}{5}$
```

Når vi skal addere eller subtrahere med brøk må vi først utvide alle brøkene til fellesnevner. Fellesnevneren er et tall som alle nevnerene går opp i (fellesnevneren kan divideres på alle de opprinnelige nevnerne). Vi vil alltid finne en fellesnevner ved å multiplisere alle nevnerene, men som oftest vil vi kunne finne enklere fellesnevnere enn det. Vi kommer tilbake til teknikker for å finne fellesnevner når vi arbeider med brøk og algebra.

```{admonition} Addisjon og subtraksjon av brøk
:class: eksempel, dropdown

$$
\begin{align}
\frac{5}{12}+\frac{1}{4}-\frac{1}{3} &=\frac{5}{12}+\frac{3}{12}-\frac{4}{12} \\
&= \frac{5+3-4}{12} \\ 
&=\frac{4}{12} \\
&= \frac{1}{3}
\end{align}
$$
```

Multiplikasjon med brøk utføres ved å multiplisere teller med teller og nevner med nevner. Et viktig spesialtilfelle er multiplikasjon av brøk med heltall. Det kan vi utføre ved å gjøre om heltallet til en brøk med $1$ i nevneren. Da ser vi at det effektivt sett blir det samme som å multiplisere heltallet inn i telleren.

```{admonition} Multiplikasjon med brøk
:class: eksempel, dropdown

**Multiplikasjon av to brøker**

$
\frac{1}{4}\cdot\frac{2}{3} =\frac{1\cdot 2}{4\cdot 3}=\frac{2}{12}=\frac{1}{6}
$

**Multiplikasjon av brøk med heltall**

$
5\cdot\frac{2}{3}=\frac{5}{1}\cdot\frac{2}{3}=\frac{5\cdot 2}{1\cdot 3}=\frac{10}{3}
$

```

Divisjon med brøk utføres ved å multiplisere med den omvendte brøken. At det må være slik ser vi ved å skrive det som en brudden brøk (brøk med brøk) og utvide.

```{admonition} Divisjon med brøk
:class: eksempel, dropdown

$$
\frac{1}{4} : \frac{2}{3}=\frac{1}{4}\cdot \frac{3}{2}=\frac{1\cdot 3}{4\cdot 2}= \frac{3}{8}
$$

**Forklaring med brudden brøk**

$$
\frac{\frac{1}{4}}{\frac{2}{3}}=\frac{\frac{1}{4}\cdot 4\cdot 3}{\frac{2}{3}\cdot 4\cdot 3}=\frac{1\cdot 3}{4\cdot 2}=\frac{1}{4}\cdot \frac{3}{2}
$$
```

Helt til slutt nevner vi at en andel skrevet som brøk kan gjøres om til henholdsvis desimaltall og prosentandel.

```{admonition} Brøk, desimaltall og prosent
:class: eksempel, dropdown

$\frac{1}{5}=0,20=20 \%$
```
