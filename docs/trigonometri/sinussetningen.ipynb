# Sinussetningen

Sinussetningen er en direkte konsekvens av arealsetningen. Vi husker at vi med arealsetningen kan finne arealet på tre måter.

$$\frac{1}{2}bc\sin A = \frac{1}{2}ac\sin B = \frac{1}{2}ab\sin C$$

Vi deler med $\frac{1}{2}abc$ i hvert uttrykk. Da får vi:

$$\frac{\sin A}{A}=\frac{\sin B}{B}=\frac{\sin C}{C}$$

Dette er sinussetningen.

```{admonition} Sinussetningen
For alle trekanter gjelder det at forholdet mellom sinus til en vinkel og lengden av motstående side er konstant.

$$\frac{\sin A}{A}=\frac{\sin B}{B}=\frac{\sin C}{C}$$

Vi kan også skrive sinussetningen *oppned*:

$$\frac{A}{\sin A}=\frac{B}{\sin B}=\frac{C}{\sin C}$$

```

Sinussetningen viser at de tre forholdene er lik hverandre. Når vi skal bruke sinussetningen velger vi to av forholdene og setter de lik hverandre.

Dersom vi skal bruke sinussetningen til å finne en ukjent vinkel må vi være oppmerksom på at det kan være to vinkler (supplementvinkler) som passer med opplysningene i oppgaven.

`````{admonition} Eksempel: Bruk av sinussetningen
:class: eksempel

I $\bigtriangleup ABC$ er $a=6$, $b=8$, og $\angle A= 35^{\circ}$. Finn $\angle B$.

Vi starter med å tegne en hjelpefigur. Vi starter med å tegne $b=8$ og $\angle A = 35^{\circ}$. Så setter vi en passer i $C$ med radius 6 og finner to mulige punkt $B_1$ og $B_2$.

```{figure} ./bilder/sinussetningen.png
---
scale: 40%
---
```

Så bruker vi sinussetningen

$
\begin{align}
\frac{\sin A}{a} &= \frac{\sin B}{b} \\
\sin B &= \frac{b\cdot \sin A}{a} = \frac{8\sin 35^{\circ}}{6}\approx 0,76 \\
\angle B &= \sin^{-1} 0,76 \approx 49,5^{\circ}
\end{align}
$

Fra figuren ser vi at det her må være $\angle B_2$. Det vil si at de to mulighetene for $\angle B$ er:

$
\angle B_1 = 180^{\circ} - 49,5^{\circ} = 130,5^{\circ} \\
\angle B_2 = 49,5^{\circ}
$

`````

Hvis vi ikke tegner skikkelig hjelpefigur som viser at det er mulig med to løsninger, bør vi til slutt gjøre en vurdering av om løsningene stemmer. Det ser vi i neste eksempel.


`````{admonition} Eksempel: Bruk av sinussetningen
:class: eksempel

I $\bigtriangleup ABC$ er $a=6$, $b=8$, og $\angle A= 50^{\circ}$. Finn $\angle B$.

Vi starter med å tegne en hjelpefigur. Vi starter med å tegne $b=8$ og $\angle A = 35^{\circ}$. Så setter vi en passer i $C$ med radius 6 og finner to mulige punkt $B_1$ og $B_2$.

```{figure} ./bilder/sinussetningen.png
---
scale: 40%
---
```

Så bruker vi sinussetningen

$
\begin{align}
\frac{\sin A}{a} &= \frac{\sin B}{b} \\
\sin B &= \frac{b\cdot \sin A}{a} = \frac{8\sin 35^{\circ}}{6}\approx 0,76 \\
\angle B &= \sin^{-1} 0,76 \approx 49,5^{\circ}
$

Fra figuren ser vi at det her må være $\angle B_2$. Det vil si at de to mulighetene for $\angle B$ er:

$
\angle B_1 = 180^{\circ} - 49,5^{\circ} = 130,5^{\circ} \\
\angle B_2 = 49,5^{\circ}
$

`````
