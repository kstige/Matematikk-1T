# Formler

Formler blir ofte brukt i realfagene og faller naturlig inn i arbeid med algebra og likninger.

```{admonition} Formel
En formel er en sammenheng mellom to eller flere størrelser.
```

For å beherske bruken av formler skal vi se på følgende steg

 * sette tall inn i formler
 * gjøre om tekst til formel
 * løse likninger fra formler
 * snu om på formler
 * (formler i programmering)

## Sette inn tall i formler
Den mest grunnleggende bruken av en formel er å sette inn tall i den for å regne ut en bestemt størrelse.

```{admonition} Eksempel: Areal av en sirkel
:class: eksempel
Arealet av en sirkel er gitt ved $A=\pi r^2$. Dersom radien er 4 vil arealet bli

$$A=\pi\cdot 4^2=16\pi$$
```

## Gjøre om tekst til formel
En viktig grunn til å bruke formler i matematikk og andre realfag er at de hjelper oss å uttrykke sammenhenger på en kompakt måte. I eksempelet over brukte vi formelen $A=\pi r^2$ for arealet av en sirkel. Skulle vi uttryk den ved tekst kunne vi sagt

*Arealet er $\pi$ multiplisert med kvadratet av radien til sirkelen.*

Eller om vi skulle brukt enda litt mer av det matematiske språket:

*Arealet er proporsjonalt med kvadratet av radien til sirkelen, med $\pi$ som proporsjonalitetskonstant.*

Slike formuleringer er viktig for å forstå innholdet i formler og for å komme fram til nye formler, men de er lite hensiktsmessige i praktiske utregninger. Der er formlene mer kompakte og hendige. Vi skal allikevel se på noen viktige matematiske begrep som kan dukke opp i forbindelse med formler.

| **Informasjon**                     | **Formel**      |
|-------------------------------------|-----------------|
| Produktet av $x$ og $y$             | $x\cdot y$      |
| Forholdet mellom $x$ og $y$         | $\frac{x}{y}$   |
| Kvadratet av $x$                    | $x^2$           |
| $y$ er proporsjonal med $x$         | $y=kx$          |
| $y$ er omvendt proporsjonal med $x$ | $y=\frac{k}{x}$ |

De to siste er ikke nødvendigvis helt kjent for alle. De sier noe om en sammenheng mellom størrelsene $y$ og $x$. Størrelsen $k$ er et reellt tall ($k\in\mathbf{R}$) og kalles proporsjonalitetskonstanten.

```{admonition} Eksempel: Timelønn
:class: eksempel
Dersom du tjener 200 kr i timen og jobber $x$ timer, så vil lønnen $L$ være gitt ved

$$L=200x$$

Her sier vi at $L$ er proporsjonal med $x$ med proporsjonalitetskonstant på 200 kr.

Når to størrelser er proporsjonale, så vokser de i samme takt. Dersom den ene størrelsen dobles, så dobles den andre. Dersom den ene halveres, så halveres den andre.
```

```{admonition} Eksempel: Akkordlønn
:class: eksempel
Du får 20000 kr for å male et hus. Du bruker $x$ timer på jobben. Timelønnen $T$ er gitt ved

$$T=\frac{20000}{x}$$

Vi sier at $T$ er omvendt proporsjonal med $x$ med proporsjonalitetskonstant på 20000 kr.

Når to størrelser er omvendt proporsjonale så endres de i hver sin retning. Dersom den ene størrelsen dobles, så vil den andre halveres.
```

```{admonition} Eksempel: Formel fra tekst
:class: eksempel

En av verdens mest kjente formler sier at energien $E$ til en gjenstand med masse $m$ er proporsjonal med kvadratet av lysfarten $c$. Proporsjonalitetskonstanten er massen til gjenstanden. Sammenhengen kan uttrykkes som

$$E = mc^2$$
```

## Løse likning fra formel
Vi startet å se på formler med å sette inn en verdi for radien i formelen $A=\pi r^2$. Da kunne vi regne direkte ut hva arealet til sirkelen var. Men hva gjør vi om vi vet arealet og trenger å finne radien?

```{admonition} Eksempel: Finne radius fra areal
:class: eksempel
Arealet av en sirkel er 20. Finn radien til sirkelen.

Vi starter med å sette inn de kjente tallene i formelen og får $20 = \pi r^2$. Her ser vi at vi ender opp med ei likning siden den ukjente størrelsen ikke står alene på den ene siden av likhetstegnet. Vi deler derfor på $\pi$ på begge sider av likhetstegnet før vi tar kvadratroten og får

$$r=\sqrt{\frac{20}{\pi}}$$
```

Vi tar også et litt mer sammensatt eksempel.

```{admonition} Eksempel: Strekning
:class: eksempel
I fysikk har vi en sammenheng mellom strekning $s$ (m), startfart $v_0$ (m/s), akselerasjon $a$ ($\text{m/s^2}$) og tid $t$ (s) som sier at

$$s=v_0t+\frac{1}{2}at^2$$

Vi skal finne akselerasjonen dersom vi får vite at strekningen er 96 m på 4 s med ei startfart på 10 m/s. Vi setter inn alle kjente tall og får 

$
\begin{align}
96 &= 10\cdot 4+\frac{1}{2}a\cdot 4^2 \\
96 &= 40+8a \\
8a &= 56 \\
a &= \frac{56}{8} \\
a &= 7
\end{align}
$

Det betyr at akselerasjonen er $7 \text{m/s^2}.

Oppgava kan også løses i CAS ved å skrive inn formelen, sette inn verdier for de kjente størrelsene, markere linja og trykke på knappen for eksakt løsning eller numerisk løsming.
```
