Noe tilsvarende kan brukes for å utføre divisjon med polynomuttrykk. Det ønsker vi å gjøre for å
 * forenkle brøkuttrykk av typen $\frac{x^3+6x^2+11x+6}{x+2}$
 * faktorisere polynomuttrykk av tredjegrad og oppover
 * løse likninger med polynomer

Siden brøk forkortes ved å stryke felles faktor, faktorisering gjøres ved hjelp av nullpunkt og alle likninger kan omformes til å handle om å finne nullpunkt, så vil de tre punktene over egentlig gå ut på det samme.

La oss se på tredjegradsuttrykket $P(x)=x^3+6x^2+11x+6$ og undersøke om -2 er et nullpunkt i uttrykket.

$$P(-2)=(-2)^3+6\cdot (-2)^2+11\cdot(-2)+6=-8+24-22+6=0$$

Vi ser at -2 er et nullpunkt og vet derfor at $x+2$ må være en faktor i $P(x)$. Det vil si at vi kan skrive $P(x)=(x+2)\cdot f(x)$, der $f(x)$ er det uttrykket vi står igjen med når vi har trukket ut faktoren $x+2$. Problemet vårt blir å finne ut hva $f(x)$ er.

$$\frac{P(x)}{x+2}=\frac{(x+2)\cdot f(x)}{x+2}=f(x)$$

Vi ser at vi kan finne $f(x)$ ved å dele $P(x)$ på $x+2$. Det kan gjøres ved hjelp av det vi kaller polynomdivisjon, og er en divisjonsalgoritme (framgangsmåte) som likner mye på divisjon med tall som vi startet med på denne siden. Vi viser algoritmen med et eksempel.

`````{admonition} Eksempel: Polynomdivisjon
Det første vi gjør er å undersøke hva vi må multiplisere $x+2$ med for å få et ledd med $x^3$. Det er ${\color{red}x^2}$, så vi skriver opp det først i svaret. Så multipliserer vi det med $x+2$ og får ${\color{red}x^3+2x^2}$ som vi skriver opp på neste linje helt til venstre. Dette trekker vi fra det opprinnelige uttrykket slik at vi står igjen med $4x^2+11x+6$.

Så gjør vi det samme på nytt og undersøker hva vi må multiplisere $x+2$ med for å få et ledd med $4x^2$. Det er ${\color{blue}4x}$, så det blir neste ledd i svaret. Videre multipliserer vi det med $x+2$ og får ${\color{blue}4x^2+8x}$. Dette skriver vi på neste linje og trekker fra. Da ender vi opp med $3x+6$.

Så gjør vi det samme på nytt og undersøker hva vi må multiplisere $x+2$ med for å få et ledd med $3x$. Det er ${\color{green}3}$, så det blir neste ledd i svaret. Videre multipliserer vi det med $x+2$ og får ${\color{green}3x+6}$. Dette skriver vi på neste linje og trekker fra. Da ender vi opp med $0$.

Siden vi ender opp med $0$ går divisjonen opp. Vi har altså ingen rest.

Vi kan da skrive opp at $P(x)=x^3+6x^2+11x+6=(x+2)(x^2+4x+3)$

`````

Som nevnt kan vi utnytte polynomdivisjon ved både faktorisering og løsning av likninger. Det ser vi i neste eksempel.

```{admonition} Eksempel: Løse tredjegradslikning
:class: eksempel
Vi skal løse tredjegradslikninga $x^3+6x^2+11x+6=0$. Vi kjenner igjen trejegradspolynomet på venstre side fra forrige eksempel. Der så vi at vi ved polynomdivisjon kunne komme fram til at $x^3+6x^2+11x+6=(x+2)(x^2+4x+3)$ Det vil si at vi har fått startet faktoriseringen. Produktsetningen sier at et uttrykk er lik null dersom en av faktorene er lik null. Her får vi altså at $x+3=0$ eller $x^2+4x+3$. Den første likningen er enkel og gir $x=-2$, mens vi på den andre må bruke abc-formelen. Den gir oss at $x=-3 \vee x=-1$.

Løsningene av likningen er altså $L=\{-3,-2,-1\}.

Fullstendig faktorisering av uttrykket blir $x^3+6x^2+11x+6=(x+3)(x+2)(x+1)$.
```

Det vi har sett over kan oppsummeres i noen veldig viktige sammenhengerc.

```{admonition} Nullpunkt - polynomdivisjon - faktorisering - likning
Vi lar $P(x)$ være et polynom og $k$ være et reelt tall. Da vil følgende påstander være ekvivalente. 

 * Polynomdivisjonen $P(x):x-k$ går opp
 * $P(k)=0$ ($k$ er et nullpunkt for $P(x)$)
 * Likninga $P(x)=0$ har $x=k$ som en løsning
 * $(x-k)$ er en faktor i $P(x)$

At påstandene er ekvivalente vil si at dersom vi vet at en stemmer, så må alle stemme. Og dersom vi vet at en ikke stemmer, så kan ingen av dem stemme.
```

```{admonition} Eksempel
:class: eksempel
Et polynom er gitt ved $P(x)=x^3-3x^2-6x+8$. Forklar at polynomdivisjonen $P(x):x-1$ går opp uten å utføre divisjonen.

Vi ser at $P(1)=1^3-3\cdot 1^2-6\cdot 1 + 8 = 0$. Det vil si at $x=1$ er et nullpunkt for $P(x)$ og at $P(x)$ vil inneholde faktoren $x-1$. Da må polynomdivisjonen gå opp.
```

Tilsvarende kan vi si noe om når polynomdivisjonen ikke går opp.

```{admonition} Når polynomdivisjonen ikke går opp
Vi lar $P(x)$ være et polynom og $k$ være et reelt tall.

Dersom polynomdivisjonen $P(x):x-k$ ikke går opp, så vil resten etter divisjonen være gitt ved $r=P(k)$. 

Da vil $(x-k)$ ikke være en faktor i $P(x)$.
```

`````{admonition} Eksempel: Polynomdivisjon som ikke går opp
:class: eksempel
Vi skal utføre polynomdivisjonen $P(x):x+2$, der $P(x)=x^3-2x+6$.

```{figure} ./bilder/polynomdivisjon2.png
---
scale: 50%
---
```

Vi ser over at vi får en rest ${\color{orange} 2}$. Det vil si at den siste delen av divisjonen ikke går opp og vi får $\displaystyle{\color{orange}\frac{2}{x+2}}$ igjen på slutten av svaret.

Vi ser da at $P(-2)=(-2)^3-2\cdot(-2)+6=-8+4+6=2$ som stemmer med at resten blir 2.

Dersom vi ser på polynomet $f(x)=P(x)-2$ vil imidlertid divisjonen gå opp siden vi da har at $f(-2)=P(-2)-2=2-2=0$. Prøv selv og sjekk at dette stemmer.

*Merk også at vi i utregningen over er nøye på å holde ledd med samme potens av $x$ under hverandre. Det gir bedre oversikt i polynomdivisjonen. Husk at vi kun kan trekke ledd med $x^2$ fra andre ledd med $x^2$. Så selv om $P(x)$ ikke inneholder noen $x^2$-ledd, så setter vi av plass til det. Det er som at det står $0\cdot x^2$ i første linje.*
`````
