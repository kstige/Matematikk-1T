# Lineære likninger

Den enkleste formen for likninger vi arbeider med er lineære likninger.

```{admonition} Lineære likninger
At ei likning er lineære vil si at den ukjente størrelsen kun kommer i første potens.

Eksempler:
 * $3x + 5 = 2$
 * $4x - 2 = 2x + 4$
 * $\frac{2x}{3} + 1 = \frac{x}{2}+\frac{2x+1}{3}$

Lineære likninger kan ha ingen, en eller uendelig mange løsninger. 
```

Vi har en hovedregel for løsning av lineære (og forsåvidt andre) likninger.

```{admonition} Løsning av likninger
Å løse en likning går ut på å finne verdien(e) som gjør at venstre side er lik høyre side i likningen. En vanlig metode er å omforme likningen slik at den ukjente størrelsen er alene på den ene siden av likhetstegnet. Vi gjør dette ved å utføre samme regneoperasjon på begge sider av likhetstegnet.
```

Vi starter med å se på noen enkle lineære likninger før vi går over til lineære likninger med brøk.

```{admonition} Eksempel 1
:class: eksempel, dropdown
Vi skal løse likningen $3x-4=x +2$. Vi gjør det ved først å legge til 4 på begge sider. Deretter trekker vi fra $x$ på begge sider. Siste steg blir å dividere begge sider med 2.

$
\begin{align}
3x-4+4 &= x+2+4 \\
3x - x &= x+6-x \\
2x &= 6 \\
\frac{2x}{2}&=\frac{6}{2} \\
x &= 3
\end{align}
$
```

```{admonition} Eksempel 2: Likning uten løsning
:class: eksempel, dropdown
Vi skal løse likningen $3x-4=3x+2$. Vi starter med å legge til 4 på begge sider. Deretter trekker vi fra $3x$ på begge sider.

$
\begin{align}
3x-4+4 &= 3x+2+4 \\
3x - 3x &= 3x+6-3x \\
0 &= 6 \\
\end{align}
$

Her ender vi opp med noe som helt tydelig ikke stemmer. Det vil si at likningen ikke har løsning. Senere i faget skal vi se at dette har en sammenheng med at de rette linjene vi får dersom vi tegner venstre og høyre side inn i et koordinatsystem er parallelle og aldri vil skjære hverandre.

Når ei likning ikke har løsning skriver vi gjerne $L=\emptyset$ som betyr at løsningsmengden til likningen er den tomme mengden $\emptyset$.
```

```{admonition} Eksempel 3: Likning med uendelig mange løsninger
:class: eksempel dropdown
Vi skal løse likningen $2(x+3)+x=3(x+1)+3$. Vi starter med å multiplisere ut parentesene og rydde på hver side av likhetstegnet slik at vi får $3x+6=3x+6$. Her ser vi at sidene er helt lik hverandre. Det vil si at likningen er en identitet. Da vet vi at likningen stemmer for alle verdier av $x$. Den har altså uendelig mange løsninger. Hadde vi prøvd å løse likningen videre ville vi endt opp med $0=0$. Det viser også at likningen er en identitet.
```

```{admonition} Eksempel 4: Likning med brøk
:class: eksempel, dropdown
Vi skal løse likningen $\displaystyle{\frac{x}{3}+2=\frac{x}{2}-2}$.

Man kan velge forskjellige første steg ved likning med brøk, men som oftest er det enkleste å starte med og kvitte seg med brøkene. Det gjør vi ved å multiplisere begge sider av likningen med fellesnevner.

Fellesnevneren til 2 og 3 er 6, så vi multipliserer begge sider med 6. Vær oppmerksom på at alle selvstendige ledd på hver side skal multipliseres med fellesnevneren. Vi vil ende opp med ei "normal" lineær likning uten brøk som vi løser som eksemplene over.

$
\begin{align}
6\left(\frac{x}{3}+2\right) &= 6\left(\frac{x}{2}-2\right) \\
\frac{6x}{3}+6\cdot 2 &= \frac{6x}{2}-6\cdot 2 \\
2x+12 &= 3x-12 \\
2x-3x &=-12-12 \\
-x &= -24 \\
\frac{-x}{-1} &= \frac{-24}{-1} \\
x &= 24
\end{align}
$

Her kom vi til at $-x=-24$. Husk at $-x$ er det samme som $-1\cdot x$. Derfor dividerer vi med $-1$ for å få $x$ alene.
```

```{admonition} Eksempel 5: Vanskeligere likning med brøk
:class: eksempel, dropdown
Vi skal løse likningen $\displaystyle{1-\frac{1}{2}\left(7x-\frac{5}{3}\right) = \frac{x-1}{4}}$. For lettere å se hva fellesnevneren er, så starter vi med å multiplisere ut parentesen på venstre side slik at vi får

$1-\frac{7x}{2}+\frac{5}{6} = \frac{x-1}{4}$

Vi ser at den enkleste fellesnevneren er 12, så vi fortsetter med å multiplisere med 12. Deretter løser vi normalt videre.

$
\begin{align}
12\left(1-\frac{7x}{2}+\frac{5}{6}\right) &= 12\left(\frac{x-1}{4}\right) \\
12-\frac{12\cdot 7x}{2}+\frac{12\cdot 5}{6} &= \frac{12(x-1)}{4} \\
12 - 42x+10 &=3x-9 \\
22-42x &= 3x-3 \\
-42x-3x &=-3-22 \\
-45x &= -25 \\
\frac{-45x}{-45} &= \frac{-25}{-45} \\
x &=\frac{25}{45} \\
x &= \frac{5}{9}
\end{align}
$
```

## Sette prøve på svaret

Vi kan kontrollere svaret på en likning ved sette inn svaret på venstre og høyre side i den opprinnelige likningen.

```{admonition} Eksempel: Sette prøve på svaret
:class: eksempel, dropdown

Bruker VS og HS som forkortelse for henholdsvis venstre og høyre side.

**Eksempel 1**
Likningen var $3x-4=x +2$ og svaret var $x=3.

**VS:** $3x-4=3\cdot 3-4=9-4=5$

**HS:** $x+2 = 3+2 = 5$

Vi ser at VS=HS. $x=3$ er derfor løsning på likningen.

**Eksempel 5**
Likningen var $\displaystyle{1-\frac{1}{2}\left(7x-\frac{5}{3}\right) = \frac{x-1}{4}}$ og svaret var $\displaystyle{x=\frac{5}{9}}$.

**VS:**

$\begin{align}
1-\frac{1}{2}\left(7x-\frac{5}{3}\right) &=1-\frac{1}{2}\left(7\frac{5}{9}-\frac{5}{3}\right)=1-\frac{1}{2}\left(\frac{35}{9}-\frac{15}{9}\right) \\
&=1-\frac{1}{2}\cdot \frac{20}{9}=1-\frac{10}{9}=-\frac{1}{9}
\end{align}
$

**HS:**

$\displaystyle{\frac{x-1}{4}=\frac{\frac{5}{9}-1}{4}=\frac{9\left(\frac{5}{9}-1\right)}{9\cdot 4}=\frac{5-9}{36}=\frac{-4}{36}=-\frac{1}{9}}$

Vi ser at VS=HS. $\displaystyle{x=\frac{5}{9}}$ er derfor løsning på likningen.

Det siste eksempelet involverer en del brøkregning som ikke er forklart. Gå gjerne nøye gjennom hvert steg for å forsikre deg om at du skjønner hva som skjer.
```
