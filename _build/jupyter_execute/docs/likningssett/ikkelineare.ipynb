# Ikke-lineære likningssett

Et ikke-lineært likningssett er et likningssett der en eller flere av de ukjente variablene er av grad to eller større. Vi kan også ha såkalte kryssledd der de ukjente størrelsene er multiplisert med hverandre.

For å løse ikke-lineære likningssett bruker vi stort sett akkurat de samme metodene som vi har brukt på lineære likningssett. Forskjellen kommer i antall løsninger og mer kompleks algebra/likningsløsning dersom vi løser for hånd. Addisjonsmetoden vil sjelden være hensiktsmessig ved ikke-lineære likningssett, og vi må bruke andre funksjoner i programmering (de går vi ikke inn på her).

Videre skal se på likningssettet

$$
x-y=-2 \\
2x^2-xy=8
$$

## Innsettingsmetoden

Vi starter med å løse den for hånd med innsettingsmetoden.

```{admonition} Innsettingsmetoden
:class: eksempel

Vi starter med å bruke den første likningen til å finne at $y=x+2$. Det setter vi inn i den andre likningen.

$$
2x^2-x(x+2)=8\\
2x^2-x^2-2x-8=0\\
x^2-2x-8=0
$$

Denne likningen løser vi greit med abc-formelen og får at $x=-2$ og $x=4$. Disse setter vi inn igjen i den første likningen for å finne $y$:

 * $x=-2$ gir $y=-2+2=0$
 * $x=4$ gir $y=4+2=6$

Vi har altså to løsninger som vi kan skrive som $L=\{(-2,0), (4, 6)\}$.
```

De fleste ikke-lineære likningssett vil være vanskelig/umulig å løse for hånd. Derfor må vi ta i bruk digitale metoder.
