# Tallmengder og symboler

I matematikk arbeider vi mye med tall og ofte vil vi være interessert i å se på ulike grupper av tall. Vi starter derfor med å definere noen viktige tallmengder som vi kommer til å arbeide med. En tallmengde er en samling av tall som kan inneholde et endelig eller uendelig antall elementer.

Den første tallmengden vi lærer oss er de naturlige tallene, eller telletallene. Videre kan vi utvide dem ved å legge til negative heltall. Det neste naturlige steget er å lage brøker, såkalte rasjonale tall. Når vi i tillegg tar med alle de irrasjonale tallene som ikke kan skrives som brøker, så har vi fylt opp hele tallinja av reelle tall.

`````{admonition} Noen viktige tallmengder

 * $\mathbf{N}$: Naturlige tall - Alle positive tall - $\{1,2,3,\ldots\}$
 * $\mathbf{Z}$: Hele tall - Alle hele tall - $\{\ldots, -2, -1, 0, 1, 2,\ldots\}$
 * $\mathbf{Q}$: Rasjonale tall - Alle tall som kan skrives som brøker av hele tall
 * Irrasjonale tall - Alle tall som ikke er rasjonale - $\pi, \sqrt{2}$
 * $\mathbf{R}$: Reelle tall - Alle tall på tallinja - Rasjonale + irrasjonale
 * $\emptyset$: Den tomme mengde - En mengde som ikke inneholder noe

Vi ser at rasjonale tall er en delmengde av de reelle tallene, de hele tallene er en delmengde av de rasjonale og at de naturlige tallene er en delmengde av de hele tallene. Dette kan vi illustrere med figuren under.

`````

## Lister og intervaller

Ofte vil svarene på oppgaver være konkrete lister av tall, eller intervaller på tallinjene. Da trenger vi symboler for å spesifisere disse.

```{admonition} Lister og intervaller

 * {} - brukes for lister av tall
 * [] - brukes for intervaller der endepunkt er med
 * $\langle \rangle$ - brukes for intervaller der endepunktene ikke er med
 * $\setminus$ - bortsett fra
 * $\cup$ - kalles union i matematikken og betyr omtrent det samme som *og*

For to tall $a$ og $b$, der $a\leq b$, så er

 * $\{a,b\}$ - mengden av tallene $a$ og $b$
 * $[a,b]$ - mengden av tallene fra og med $a$ til og med $b$ ($a\leq x\leq b$)
 * $\langle a,b\rangle$ - mengden av tallene mellom $a$ og $b$ ($a<x<b$)

$[a,b]$ kalles et lukket intervall, mens $\langle a, b \rangle$ kalles et åpent intervall. Vi kan også lage halvåpne intervall $[a,b\rangle$ og $\langle a, b].

For å skrive at $x$ er et element i en mengde $M$, så skriver vi $x\in M$.

For å skrive at $x$ ikke er et element i en mengde $M$, så skriver vi $x\notin M$.

Vi kan også kombinere ulike intervaller eller fjerne enkelte verdier

 * $\mathbf{Q}\cup\{\pi\}$ - mengden av rasjonale tall og pi.
 * $\mathbf{R} \setminus \mathbf{Z}$ - mengden av reelle tall som ikke er heltall

Dersom et intervall går til uendelig kan vi bruke tegnene $\infty$ eller piler $\leftarrow$ og $\rightarrow$.
```

Vi kommer til å bruke symbolene over når vi skal skrive svar på likninger og ulikheter, når vi skal spesifisere intervaller for definisjonsmengde og verdimengde for funksjoner og ved definisjon av ulike sammenhenger og mønster. Alt det her kommer vi tilbake til senere, men vi gir et kort eksempel her.

```{admonition} Partall og oddetall
:class: eksempel

Partall er tall som kan skrives på formen $2n$ der $n\in\mathbf{N}$.

Oddetall er tall som kan skrives på formen $2n-1$ der $n\in\mathbf{N}$.

Om vi ønsker å ta med de negative par- og oddetallene, så kan vi bytte ut $\mathbf{N}$ med $\mathbf{Z}$.
```

```{admonition} Sammensatt intervall på tallinja
:class: eksempel

$x\in [2, 4\rangle\cup[6,\rightarrow\rangle$

$x$ er et element i mengden av tall fra og med 2, til men ikke med 4 og fra og med 6 til uendelig.
```

## Absoluttverdi

Alle reelle tall består av et fortegn (+ eller -) og en tallverdi. Denne tallverdien kalles absoluttverdien til tallet. Absoluttverdien til tallet kan vi si at er verdien til tallet når fortegnet er tatt vekk. Vi bruker || for å skrive absoluttverdi. For eksempel vil absoluttverdien til -3 være gitt ved $|-3|=3$.

```{admonition} Absoluttverdi
Vi definerer absoluttverdien til tallet $a$ som 

$$
|a|=\begin{cases}
    a, & \text{hvis $a>0$}\\
    -a, & \text{hvis $a<0$}
  \end{cases}
$$
```

Absoluttverdier er nyttige når vi jobber med avstander og vi møter dem når vi skal definere kvadratrøtter. De kan også brukes for å skrive enkelte intervaller på en mer kompakt måte. $-2<a<2$ kan også skrives som $|a|<2$.

## Implikasjon og ekvivalens

I mange sammenhenger (spesielt i bevis og utledninger) vil det være nyttig å se på om en påstand medfører (impliserer) en annen påstand. Et ikke-matematisk eksempel er:

*Per bor i Molde* **medfører at** *Per bor i Norge*.

Vi kan imidlertid ikke snu om på resonnementet. Dersom vi vet at Per bor i Norge, så kan vi ikke være sikker på at han bor i Molde. Det her er et eksempel på det vi i matematikken kaller en implikasjon.

```{admonition} Implikasjon
Dersom det at påstanden $p$ er sann medfører at påstanden $q$ er sann, så sier vi at $p$ impliserer $q$ og skriver

$$p\Rightarrow q$$
```

```{admonition} Eksempel: Implikasjon
:class: eksempel
$x=5 \Rightarrow x^2=25$

$x>5 \Rightarrow x>4$

$x\in[4,5] \Rightarrow x\in [3,6]$
```

I en del tilfeller, så vil implikasjonen gjelde begge veier. Da har vi det vi kaller en ekvivalens. Et ikke-matematisk eksempel er:

*Per er eldre enn Kari* **er ekvivalent med at** *Kari er yngre enn Per*.

```{admonition} Ekvivalens
Dersom det at påstand $p$ er sann medfører at påstand $q$ er sann og omvendt, så sier vi at $p$ er ekvivalent med $q$ og skriver

$$p \Leftrightarrow q$$
```

Selv om vi ikke tenker over det, så er ekvivalens viktig når vi løser likninger. Dersom alt vi gjør når vi løser likningen fører til at neste linje er ekvivalent med den forrige, så er vi sikker på at det endelige svaret stemmer med den opprinnelige likningen.

```{admonition} Eksempel: Ekvivalens
:class: eksempel

$x=4 \Leftrightarrow 3x=12$

$x^2=4 \Leftrightarrow x=2 \text{ eller } x=-2$

$x+4 = 8 \Leftrightarrow x=4$
```
