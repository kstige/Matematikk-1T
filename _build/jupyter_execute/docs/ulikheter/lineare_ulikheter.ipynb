# Lineære ulikheter

En ulikhet er på mange måter det samme som ei likning og vi vil i de fleste tilfeller ta utgangspunkt i de samme løsningsmetodene som vi bruker for likninger. Men der vi ei likning ønsker å bestemme den ukjente slik at sidene blir like store, så vil vi i en ulikhet avgjøre hvilke verdier av den ukjente som oppfyller et annet type krav.

Et eksempel på en ulikhet er $3x+2<5$. Den likner veldig på likningen $3x+2=5$. I likningen ønsker vi å komme fram til verdien for $x$ som gjør at $3x+2$ blir $5$. I ulikheten ønsker vi å komme fram til alle verdiene av $x$ som gjør at $3x+2$ blir mindre enn $5$.

Som nevnt vil vi bruke mange av de samme løsningsmetodene som vi allerede har lært oss for likninger. Det er imidlertid en viktig forskjell å legge merke til.

```{admonition} Løsning av ulikheter
For å løse en ulikhet kan vi gjøre det samme på begge sider av ulikhetstegnet.

Om vi multipliserer eller dividerer med et negativt tall må vi snu ulikhetstegnet.
```

For å forklare hvorfor vi må snu ulikhetstegnet er det enklest å se på et talleksempel.

```{admonition} Eksempel: snu ulikhetstegnet
:class: eksempel, dropdown

En ulikhet vi vet at stemmer er at $5<10$. Hva skjer dersom vi multipliserer begge sider med $-1$?

$$
5<10\\
5(-1)<10(-1)\\
-5<-10
$$

Det her ser vi at ikke stemmer siden vi vet at $-5>-10$. Feilen oppstår når vi multipliserer med et negativt tall uten å snu ulikhetstegnet. Grunnen til at det blir feil er at positive tall er størst når de har stor absoluttverdi, mens negative tall er størst når absoluttverdien er liten. Når vi multipliserer (eller dividerer) med et negativt tall endrer vi fortegnet og må samtidig ta hensyn til at vi "endrer rekkefølgen" på hva som er størst og minst. Det riktige blir derfor

$$
5<10\\
5(-1)>10(-1)\\
-5>-10
$$
```

Vi skal videre se på noen eksempler med lineære ulikheter løst for hånd.

## Grafisk løsning av ulikheter

Den siste metoden vi skal se på for å løse ulikheter er grafisk.

`````{admonition} Grafisk løsning av ulikhet
:class: eksempel

Vi skal løse ulikheten $2x+1<-3x+6$ grafisk. Det første vi gjør er å skrive inn hver side av ulikheten som en funksjon i GeoGebra. Vi skriver inn $f(x)=2x+1$ og $g(x)=-3x+6$. Så bruker vi "skjæring mellom to objekt" til å finne skjæringspunktet $A=(1,3)$. Vi ser at funksjonen $f(x)$ ligger under $g(x)$ til venstre for $x=1$. Det vil si at $f(x)<g(x)$ i dette området.

Løsningen av ulikheten blir derfor $x<1$. 

For å se det litt mer visuelt kan vi også skrive inn $f(x)<g(x)$ for å få markert det blå området som løser ulikheten.
`````
