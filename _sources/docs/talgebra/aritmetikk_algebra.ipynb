# Aritmetikk og algebra

Aritmetikken er læren om tallenes egenskaper og omhandler metodene vi bruker til å regne med tall. Algebraen tar for seg generalisering av aritmetikken til bokstavuttrykk. I aritmetikken kan vi se på uttrykk som $4+4=2\cdot 4$, mens vi i algebraen arbeider med symboler og kan se på $x+x=2x$. Her skal vi se på en del av de grunnleggende regnereglene vi bruker når både til tallregning og bokstavuttrykk.



## Regnerekkefølge

De fire grunnleggende regneoperasjonene er addisjon, subtraksjon, multiplikasjon og divisjon. I tillegg kommer vi til å ha bruk for potenser og røtter (kvadratrøtter).

Når vi skal utføre et sammensatt regnestykke, så benytter vi oss av denne regnerekkefølgen

```{admonition} Regnerekkefølgen

 1) Parenteser
 2) Potenser og røtter
 3) Multiplikasjon og divisjon
 4) Addisjon og subtraksjon

```

Videre har vi en del egenskaper eller regneregler som gjelder både ved tallregning og bokstavregning. De fleste bør være kjente, men her formulerer vi dem presist.

```{admonition} Regneregler
Vi lar $a, b \in \mathbf{R}$.

**Kommutativ lov**
Addisjon og multiplikasjon kan utføres i den rekkefølgen vi ønsker. Gjelder ikke for subtraksjon og divisjon.
 1) $a+b=b+a$
 2) $a\cdot b = b\cdot a$

**Assosiativ lov**
Vi kan addere og multiplisere tre elementer ved å velge to vilkårlige av elementene først. Gjelder ikke for subtraksjon og divisjon.
 3) $a+b+c= a + (b+c)= (a+b)+c$
 4) $a\cdot b\cdot c = a\cdot(b\cdot c)=(a\cdot b)\cdot c$

**Distributiv lov**
Handler om hvordan vi multipliserer ut parenteser. Gjelder også om vi bytter en eller begge addisjonene med subtraksjon.
 5) $a\cdot(b + c)= a\cdot b + a\cdot c$
 6) $(a+b)(c+d) = ac + ad + bc +bd$

**Negativ faktor**
Multiplikasjon av to faktorer med ulikt fortegn gir negativt svar. Multiplikasjon av to faktorer med likt fortegn gir positivt svar. Gjelder også for divisjon.
 7) $(-a)\cdot b = a(-b)=-(ab)=-a\cdot b$
 8) $(-a) \cdot (-b) = a\cdot b$

```

Vi skal ta for oss noen eksempler der vi bruker blant annet den distributive loven på algebraiske uttrykk. Det er imidlertid en viktig huskeregel for arbeid med algebraiske uttrykk som vi må ta for oss først.

```{admonition} Trekke sammen algebraiske uttrykk
Når vi skal forenkle et algebraisk uttrykk kan vi kun trekke sammen ledd med helt lik variabelsammensetning. Vi legger derfor sammen $x$ for seg, $y$ for seg, $x^2$ for seg, $xy$ for seg og så videre. Når vi legger sammen, ser vi på koeffisientene (tallene) foran variablene og adderer/subtraherer dem. Er koeffisienten lik $1$ bruker vi vanligvis ikke å skrive den (vi skriver $x$ ikke $1x$).

$x+x+y=2x+y$

$2xy-x+xy+2x=3xy+x$
```

```{admonition} Eksempel: Trekke sammen algebraiske uttrykk
:class: eksempel, dropdown

Her er fire grunnleggende eksempler på å trekke sammen algebrauttrykk. Det er brukt mange ekstra mellomregninger for å tydeliggjøre hva som blir gjort.

$2a+(3a-4a)=2a+3a-4a=a$

$
\begin{align}
2a-(3a-4a)&=2a-1\cdot(3a-4a) \\
&=2a-1\cdot 3a -1\cdot(-4a)=2a-3a+4a=3a
\end{align}
$

$4(2-a)=4\cdot 2+4\cdot(-a)=8-4a$

$
\begin{align}
(4+2a)(2-a)&=4\cdot 2+4\cdot(-a)+2a\cdot 2+2a\cdot (-a) \\
&=8-4a+4a-2a^2=-2a^2+8
\end{align}
$
```

I det sist eksempelet kom vi bort i uttrykket $a^2$. Det er et eksempel på en potens som vi skal se nærmere på i det neste avsnittet.

## Potenser
En potens er en gjentatt multiplikasjon.

```{admonition} Definisjon av potens
Dersom $a\in\mathbf{R}$ og $n\in\mathbf{N}$, så betyr $a^n$ at vi skal multiplisere $a$ med seg selv $n$ ganger. $a$ kalles grunntallet, mens $n$ er eksponenten.

$$a^n=\underbrace{a\cdot a \cdots a}_{n \text{ ganger}}$$
```

All regning med potenser kan utføres fra definisjonen, men det blir mye mer effektivt dersom vi tar i bruk regnereglene for potenser. Vi beviser ikke alle regnereglene her, men i eksemplene som følger ser vi at de kommer som en ganske naturlig følge av definisjonen over.

```{admonition} Regneregler for potenser

 1) $a^n \cdot a^m = a^{n+m}$
 2) $\frac{a^n}{a^m} = a^{n-m}$
 3) $\left(a\cdot b\right)^n = a^n \cdot a^m$
 4) $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$
 5) $\left(a^n\right)^m = a^{n\cdot m}$
 6) $a^0 = 1$
 7) $a^{-n} = \frac{1}{a^n}$

```

```{admonition} Eksempel: "Bevis" for regneregel 1
:class: eksempel, dropdown
$3^4\cdot 3^5=\underbrace{\underbrace{3\cdot 3\cdot 3\cdot 3}_{4 \text{ ganger}}\cdot \underbrace{3\cdot 3\cdot 3\cdot 3\cdot 3}_{5 \text{ ganger}}}_{4+5 \text{ ganger }=9\text{ ganger }}=3^{4+5}=3^9$
```
```{admonition} Eksempel: Enkle brøkuttrykk
:class: eksempel, dropdown
$\displaystyle{\frac{4^5}{4^3}=4^{5-3}=4^2}$

$\displaystyle{\frac{5^3\cdot 5}{5^2\cdot 5^{-3}}=\frac{5^{3+1}}{5^{2+(-3)}}=\frac{5^4}{5^{-1}}=5^{4-(-1)}=5^5}$
```

```{admonition} Eksempel: "Bevis" for regneregel 6
:class: eksempel, dropdown
Vi ser på uttrykket $\frac{a^3}{a^3}$ på to måter.

$\displaystyle{\frac{a^3}{a^3}}=1$ siden vi deler et tall på seg selv.

$\displaystyle{\frac{a^3}{a^3}=a^{3-3}=a^0}$ ved regneregel 2.

Siden $\frac{a^3}{a^3}$ er lik både $1$ og $a^0$, så må $a^0=1$.
```

```{admonition} Eksempel: Sammensatte potensuttrykk
:class: eksempel, dropdown

$\displaystyle{\frac{(2x)^3\cdot x^2}{x^3}=\frac{2^3\cdot x^3 \cdot x^2}{x^3}=\frac{2^3\cdot x^(3+2)}{x^3}=2^3\cdot x^{5-3}=8x^2}$

$\displaystyle{\frac{2a^3\cdot (2a)^3}{\left(\frac{1}{2}a^2\right)^{-3}}=\frac{2a^3\cdot 2^3 a^3}{2^{(-1)(-3)}a^{2(-3)}}=\frac{2^{1+3}a^{3+3}}{2^3 a^{-6}}=2^{4-3}a^{6-(-6)}=2a^{12}}$

```
### Standardform
Når vi arbeider med tall som er veldig store, eller med tall som er veldig små kan en spesiell skrivemåte med potenser hjelpe oss.

```{admonition} Standardform
Standardform for et tall er

$a\cdot 10^n$, der $n\in \mathbf{Z}$ og $|a|\in[1,10\rangle$.

Vi skriver altså tallet med et siffer (1 eller større) før komma og multipliserer med en tierpotens for å beksrive størrelsen på tallet.
```

Standardform forklares enklest med noen eksempler.

```{admonition} Eksempel: Skrive tall på standardform
:class: eksempel, dropdown

For å skrive et tall på standardform teller vi antall plasser vi må flytte komma for å ha kun et siffer større eller lik 1 før komma. Eksponenten til tierpotensen blir antall plasser vi har flyttet komma mot venstre.

$53600000=5,36\cdot 10^7$

Her har vi flyttet komma 7 plasser mot venstre.

$0,0000000345 = 3,45\cdot 10^{-8}$

Her har vi flyttet komma 8 plasser mot høyre og får en negativ eksponent. Det er fordi $0,0000000345=\frac{3,45}{10^8}$.
```

```{admonition} Eksempel: Regne med potenser og standardform
:class: eksempel, dropdown

$\frac{160000\cdot 0,02}{0,008}=\frac{16\cdot 10^4\cdot 2\cdot 10^{-2}}{8\cdot 10^{-3}}=\frac{16\cdot 2}{8}\cdot 10^{4+(-2)-(-3)}=4\cdot 10^5$

Her har vi ikke brukt skikkelig standardform før i svaret, men vi ser hvordan regnestykket forenkles ved å bruke tierpotenser og potensreglene.
```



## Kvadratrøtter
De fleste vet nok at $\sqrt{16} = 4$, men det kan være greit å definere begrepet kvadratrot skikkelig siden vi kommer til å benytte det inn mot blant annet andregradslikninger senere.

```{admonition} Definisjon av kvadratrot
Vi lar $a$ være et positivt reelt tall. Kvadratroten av $a$ er det positive tallet som multiplisert med seg selv gir $a$.

$$\left(\sqrt{a}\right)^2=a$$
```

I definisjonen over er det to ting som er viktig å legge merke til.
 1) Vi kan kun ta kvadratroten av positive reelle tall.
 2) Kvadratroten er et positivt tall. Selv om både $(-4)^2=16$ og $4^2=16$, så er $\sqrt{16}= 4$.

Generelt kan vi skrive at $\sqrt{a^2}=|a|$, der $|a|$ er absoluttverdien til tallet $a$.

Om kvadratrøtter kan det videre være nyttig å vite følgende.

```{admonition} Viktig om kvadratrøtter
$$\sqrt{a+b}\neq \sqrt{a}+\sqrt{b}$$

$$\sqrt{a\cdot b} = \sqrt{a}\cdot \sqrt{b}$$
```

Om vi tenker oss en del om, så ser vi at den gyldige sammenhengen over minner om regneregel 3 for potenser. Den sa at $(ab)^n=a^n b^n$. Dette er ikke en tilfeldighet. Kvadratrøtter kan faktisk skrives som en potens. Det er ikke pensum i 1T, men vi nevner det her som bakgrunnsstoff.

```{admonition} Kvadratrot og potens
:class: dropdown
Dersom $a$ er et positivt reelt tall, så er

$$\displaystyle{\sqrt{a}=a^{\frac{1}{2}}}$$

Vi ser at dette må stemme siden

$$(\sqrt{a})^2=(a^{\frac{1}{2}})^2=a^{\frac{1}{2}\cdot 2}=a^1=a$$

ALle regneregler for potenser må derfor også gjelde for kvadratrøtter.
```

Helt til slutt tar vi med den generelle definisjonen på n-te røtter.

```{admonition} n-terøtter
:class: dropdown

n-teroten av et tall $a$ er det tallet som opphøyd i $n$ er lik $a$.

$$(\sqrt[n]{a})^n=a$$

Dersom $n$ er et partall, så må både $a$ og $\sqrt[n]{a}$ være positive tall.

Videre, så har vi at

$$a^{t}{n}=\sqrt[n]{a^t}=(\sqrt[n]{a})^t$$

der $t$ er et heltall, $n$ er et positivt heltall og $a$ er et positivt tall.
```
