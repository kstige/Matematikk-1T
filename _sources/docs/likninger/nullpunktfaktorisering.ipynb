# Nullpunktfaktorisering

Vi har sett hvordan uttrykk kan faktoriseres ved:

 * å trekke ut felles faktor
 * kvadratsetningene
 * fullstendig-kvadratmetoden

Vi har også sett på hvordan vi kan løse andregradslikninger på formen $ax^2+bx+c=0$. Når vi løser denne likningen finner vi nullpunktene til andregradspolynomet $ax^2+bx+c$. Vi skal her se på hvordan vi kan bruke nullpunkt til å faktorisere generelle andregradsuttrykk uten å bruke fullstendig-kvadratmetoden, men først definerer vi begrepene andregradspolynom og nullpunkt.

```{admonition} Andregradspolynom
Et uttrykk som kan skrives på formen

$P(x)=ax^2+bx+c$ der $a$, $b$, $c \in \mathbf{R}$ og $a\neq 0$

kalles et andregradspolynom.

Ordet polynom består av ordene

 * poly - mange
 * nom - ledd

og er altså et uttrykk som består av flere ledd.

Ved å legge til ledd med $x^3$ vil vi få et tredjegradspolynom og så videre.
```

```{admonition} Nullpunkt
Når vi løser en likning $P(x)=0$ finner vi nullpunktene til uttrykket $P(x)$. Nullpunktene er de $x$-verdiene som gjør at uttrykket blir lik null.

Dersom $a$ er et nullpunkt for $P(x)$, så vil $P(a)=0$.
```

```{admonition} Eksempel: Andregradspolynom og nullpunkt
:class: eksempel

$P(x)=x^2+2x-8$ er et andregradspolynom. $x=2$ og $x=3-4$ er nullpunkt til andregradspolynomet siden

$P(2)=2^2+2\cdot2-8=4+4-8=0$

$P(-4)=(-4)^2+2\cdot(-4)-8=16-8-8=0$
```

Kan vi tenke oss et faktorisert uttrykk som har de samme nullpunktene? Vi angriper problemet ved å ta en del av gangen.

Det enkleste uttrykket som har $2$ som nullpunkt er: $x-2$.

Det enkleste uttrykket som har $-4$ som nullpunkt er: $x+4$.

Hvis vi husker produktsetningen fra likningsløsningen, så vil det enkleste uttrykket som har $2$ og $-4$ som nullpunkt være produktet av disse:

$$(x-2)(x+4)$$

Hvis vi multipliserer ut dette uttrykket ender vi opp med 

$$(x-2)(x+4)=x^2+4x-2x-2\cdot 4 = x^2-2x-8$$

Det er uttrykket vi så på i eksempelet over. Har vi nullpunktene til et uttrykk kan vi altså raskt faktorisere uttrykket. Vi uttrykker dette generelt.

```{adminition} Nullpunktfaktorisering
Dersom andregradspolynomet $P(x)=ax^2+bx+c$ har nullpunktene $x_1$ og $x_2$, få svil faktorisert form av $P(x)$ være

$P(x)=ax^2+bx+c=a(x-x_1)(x-x_2)$

Om $x_1=x_2$ (dobbelt nullpunkt) vil vi få $P(x)=(x-x_1)^2$

Om $P(x)$ ikke har nullpunkt, så kan det ikke faktoriseres.

Framgangsmåten kan lett generaliseres til polynom av høyere grad.
```

Forhåpentligvis gir resultatet mening, men det kan også bevises algebraisk:

```{admonition} Bevis for nullpunktfaktorisering
:class: dropdown

Vi ser på et generelt andregradspolynom $P(x)=ax^2+bx+c$. Fra abc-formelen vet vi at løsningene er

$x_1=\frac{-b+\sqrt{b^2-4ac}{2a}}$
$x_2=\frac{-b-\sqrt{b^2-4ac}{2a}}$

Vi setter disse løsningene inn i faktoriseringen. For oversikten sin skyld setter vi $\sqrt{b^2-4ac}=d$ fram til slutten av beviset.

$\begin{align}
P(x)=ax^2+bx+c &=a(x-x_1)(x-x_2) \\
&=  a\left(x-\frac{-b+d}{2a}}\right)\left(x-\frac{-b-d}{2a}}\right) \\
&= a\left(x^2-\frac{x(-b-d)}{2a}-\frac{x(-b+d)}{2a}+\frac{(-b+d)(-b-d)}{4a^2}\right) \\
&= ax^2 +\frac{xb+xd+xb-xd}{2}+\frac{b^2-d^2}{4a} \\
&= ax^2+bx\frac{b^2-(b^2-4ac)}{4a} \\
&= ax^2+bx+c
\end{align}
$



```

Vi tar noen eksempler

```{admonition} Eksempel: Nullpunktfaktorisering
:class: eksempel
Finner nullpunktene til uttrykkene på venstre side ved abc-formelen og setter opp faktoriseringen.

$x^2-5x+6 = (x-3)(x-2)$ siden nullpunktene er $2$ og $3$.

$x^2+7x+10=(x+2)(x+5)$ siden nullpunktene er $-2$ og $-5$.

$2x^2-10x+12 = 2(x-3)(x-2)$ siden nullpunktene er $2$ og $3$. Legg merke til at vi må få med faktoren 2 som står foran andregradsleddet.

Kontroller gjerne faktoriseringen ved å multiplisere ut parentesene på høyre side, eller ved å bruke fullstendigkvadratmetoden.
```
