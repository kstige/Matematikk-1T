# Andregradslikninger

Fram til nå har vi arbeidet med lineære likninger. Det er likninger der den ukjente kun er i første potens. Skal vi bruke likninger til å løse problemstillinger innenfor andre fagfelt, så vil vi imidlertid ende opp med likninger der den ukjente er opphøyd i andre. Eksempler på slike likninger er

 1) $x^2=36$
 2) $2x^2-8=0$
 3) $x^2-12x=0$
 4) $x^2+6x+9=4$
 5) $x^2-12x+20=0$


```{admonition} Andregradslikninger
I en andregradslikning vil den ukjente størrelsen være i andre potens. Alle andregradslikninger kan skrives på formen

$$ax^2+bx+c=0$$

der $a, b, c \in \mathbf{R}$ og $a\neq 0$.
```

Dersom vi flytter alle ledd over på venstre side, så vil alle eksemplene over være på denne generelle formen. I de tre første er enten $b$ eller $c$ lik null. Vi starter med å se på disse tilfellene.

## $ax^2+c=0$ 

Vi starter med andregradslikninger der $b=0$. Da er likningen på formen $ax^2+c=0$. Slike likninger løser vi ved enklest ved å flytte $c$ over til høyre side, dividere på $a$ og så ta kvadratrot. I det første eksempelet viser vi hvordan vi løser den generelt før vi løser 1) og 2) i lista øverst på sida.

```{admonition} Eksempel: $ax^2+c=0$
:class: eksempel
$
\begin{align}
ax^2+c &= 0 \\
ax^2 &= -c \\
x^2 &= -\frac{c}{a} \\
\sqrt{x^2} &= \sqrt{-\frac{c}{a}} \\
|x| &= \sqrt{-\frac{c}{a}} \\
x &= \pm \sqrt{-\frac{c}{a}}
\end{align}
$

Her har vi brukt at $\sqrt{x^2}=|x|$ siden vi har definert kvadratroten til å være et positivt tall. 

Vi husker at vi kun kan ta kvadratroten av positive tall. Dersom vi skal kunne regne ut kvadratroten på høyre side må derfor enten $a$ eller $c$ være negativ (men ikke begge). Om $a$ og $c$ har likt fortegn vil vi få et negativt tall inne i kvadratoten og den er da ikke definert. I slike tilfeller har likningen ingen løsning. Videre vil vi ha en løsning dersom $c=0$ og to løsninger (gitt ved $\pm$ foran kvadratroten) dersom $a$ og $c$ har forskjellige fortegn.
```

Det vi ser over gjelder generelt for alle andregradslikninger.

```{admonition} Antall løsninger
En andregradslikning kan ha

 * ingen løsning
 * en løsning
 * to løsninger
```

Vi tar nå to talleksemepel

```{admonition} $x^2=36$
:class: eksempel

I denne likningen er vi klar til å ta kvadratrot med en gang

$
\begin{align}
x^2 &= 36 \\
\sqrt{x^2} &= \sqrt{36} \\
|x| &= 6 \\
x &= \pm 6
\end{align}
$

Det er også vanlig å skrive opp løsningen som $L=\{-6, 6\}$ som betyr at løsningsmengden til likningen er $\{-6, 6\}$

Vanligvis vil vi hoppe over linje 3 i mellomregningen her.
```

```{admonition} $2x^2-8=0$
:class: eksempel

$
\begin{align}
2x^2 -8 &= 0 \\
2x^2 &= 8 \\
x^2 &= 4 \\
\sqrt{x^2} &= \sqrt{4} \\
|x| &= 2 \\
x &= \pm 2
\end{align}
$

Løsningsmengden er $L=\{-2,2\}$.
```

## $ax^2+bx=0$

Den neste formen for andregradslikning vi skal se på er likninger der konstantleddet $c=0$. For å løse disse kan vi bruke noe vi kaller for produktsetningen.

```{admonition} Produktsetningen
Dersom produktet av to tall blir null, så må minst et av tallene være null.

$$x\cdot y = 0 \quad \Leftrightarrow \quad x = 0 \vee y = 0$$
```

Vi skal i første omgang bruke produktsetningen til å løse likninger på formen $ax^2+bx=0$, men den vil være veldig nyttig også når vi skal se på faktorisering av generelle andregradsuttrykk.

```{admonition} $ax^2+bx=0$
:class: eksempel

Første steg i løsningen er å se at vi har en felles faktor $x$. Vi starter med å trekke den utenfor en parentes.

$ax^2+bx=x(ax+b)=0$

Her har vi et produkt av to faktorer som skal bli null. Ved produktsetningen er det kun mulig dersom minst en av faktorene er null. Løsningen på likningen er derfor

$
\begin{align}
x=0 \qquad \vee& \qquad ax+b=0\\
x=0 \qquad \vee& \qquad ax=-b \\
x=0 \qquad \vee& \qquad x=-\frac{b}{a}
\end{align}
$

Løsningen blir: $L=\{0, -\frac{b}{a}\}$.
```

Så tar vi et talleksempel.

```{admonition} $x^2-12x=0$
:class: eksempel
Vi starter med å faktorisere ut felles faktor $x$.

$x^2-12x=x(x-12)=0$

Her har vi et produkt av to faktorer som skal bli null. Ved produktsetningen er det kun mulig dersom minst en av faktorene er null. Løsningen på likningen er derfor

$
\begin{align}
x=0 \qquad \vee& \qquad x-12=0\\
x=0 \qquad \vee& \qquad x=12 
\end{align}
$

Løsningen blir: $L=\{0, 12\}$.
```


## Generelle andregradslikninger

I en helt generell andregradslikning vil ingen av koeffisienten være lik null. Det vil derfor ikke være mulig å bruke de spesielle framgangsmåtene over for å løse likningene. Prinsippene vi har brukt er imidlertid nyttige.

Vi har sett at dersom vi kan faktorisere uttrykket, så kan vi bruke produktsetningen til å finne løsningene. Andregradsuttrykk kan faktoriseres ved å lage fullstendige kvadrat slik vi har sett på tidligere. 

```{admonition} $x^2+6x+9=4$
:class: eksempel

Vi skal løse oppgaven på to måter.

**Metode 1**
Vi ser at venstresiden av uttrykket er et fullstendig kvadrat. Vi kan derfor faktorisere det direkte slik at vi får.

$(x+3)^2=4$

Videre følger vi framgangsmåten fra likningene med $b=0$ ved å ta kvadratrot på begge sider.

$
\begin{align}
\sqrt{(x+3)^2}&=\sqrt{4} \\
|x+3|&= 2 \\
x+3 &= \pm 2 
\end{align}
$

Vi får to løsninger gitt ved

$
\begin{align}
x+3 = 2 \qquad &\vee \qquad x+3=-2 \\
x = 2-3 \qquad &\vee \qquad x=-2-3 \\
x = -1 \qquad &\vee \qquad x=-5
\end{align}
$

Løsningen blir $L=\{-5, -1\}$.

**Metode 2**
Vi bruker første kvadratsetning på venstre side og flytter $4$ over til venstre side. Videre skriver vi $4$ som et kvadrat $2^2$ slik at vi står med:

$(x+3)^2-2^2=0$

Her kan vi bruke konjugatsetningen på venstre side.

$
\begin{align}
(x+3+2)(x+3-2)&=0 \\
(x+5)(x+1) &= 0
\end{align}
$

Her har vi et produkt av to faktorer som skal bli null. Ved produktsetningen er det kun mulig dersom minst en av faktorene er null. Løsningen på likningen er derfor

$
\begin{align}
x+5=0 \qquad \vee& \qquad x+1=0\\
x=-5 \qquad \vee& \qquad x=-1 
\end{align}
$

Løsningen blir $L=\{-5, -1\}$.
```

Vi tar et eksempel til der vi må lage oss et fullstendig kvadrat. Også dette eksempelet utfører vi på to måter.

```{admonition} $x^2-12x+20=0$
:class: eksempel

**Metode 1**
Vi starter med å lage et fullstendig kvadrat på venstre side. For å få det må vi legge til og trekke fra $\left(\frac{-12}{2}\right)^2=36$.

$x^2-12x+20=\underbrace{x^2-12x+36}_{\text{Fullstendig kvadrat}}-36+20=(x-6)^2-16=0$

Videre flytter vi $-16$ over til høyre side (ved å legge til $16$ på begge sider av likhetstegnet) og følger framgangsmåten fra likningene med $b=0$ ved å ta kvadratrot på begge sider.

$
\begin{align}
\sqrt{(x-6)^2}&=\sqrt{16} \\
x-6 &= \pm 4 
\end{align}
$

Vi får to løsninger gitt ved

$
\begin{align}
x-6 = 4 \qquad &\vee \qquad x-6=-4 \\
x = 4+6 \qquad &\vee \qquad x=-4+6 \\
x = 10 \qquad &\vee \qquad x=2
\end{align}
$

Løsningen blir $L=\{2, 10\}$.

**Metode 2**
Vi starter på samme måte ved å lage et fullstendig kvadrat på venstre side. Videre skriver vi $16$ som et kvadrat $4^2$ slik at likningen vår blir

$(x-6)^2-4^2=0$

Her kan vi bruke konjugatsetningen på venstre side.

$
\begin{align}
(x-6+4)(x-6-4)&=0 \\
(x-2)(x-10) &= 0
\end{align}
$

Her har vi et produkt av to faktorer som skal bli null. Ved produktsetningen er det kun mulig dersom minst en av faktorene er null. Løsningen på likningen er derfor

$
\begin{align}
x-2=0 \qquad \vee& \qquad x-10=0\\
x=2 \qquad \vee& \qquad x=10 
\end{align}
$

Løsningen blir $L=\{2, 10\}$.
```

## abc-formel

De to metodene som er brukt i eksemplene over er nok for å løse alle andregradslikninger som kan løses. Eksemplene er imidlertid valgt med "fine tall" som gjør det lett å faktorisere. I en andregradslikning som beskriver en eller annen reell situasjon vil vi sjeldent få fine tall. Det vil derfor være praktisk med en formel som vi kan bruke direkte på alle andregradslikninger. Denne formelen kalles abc-formelen og vi gir den her før vi tar bevis og en del eksempler på bruk.

```{abc-formelen}
Løsningen av andreagradslikningen $ax^2+bx+c=0$ er gitt ved:

$$x=\frac{-b\pm\sqrt{b^2-4ac}{2a}}$$
```

```{admonition} Bevis for abc-formelen
:class: eksempel, dropdown, full-width
Beviset for abc-formelen bygger på faktoriseringsmetodene vi brukte over.

$
\begin{align}
ax^2+bx+c&=0 \quad & \\
ax^2+bx&=-c \quad & \text{Trekker fra c på begge sider.} \\
x^2+\frac{b}{a}x &=-\frac{c}{a} \quad &\text{Dividerer med a på begge sider.} \\
\underbrace{x^2+\frac{b}{a}x+\left(\frac{b}{2a}\right)^2}_{\text{Fullstendig kvadrat}} &=-\frac{c}{a}+\left(\frac{b}{2a}\right)^2 \quad &\text{Legger til }\left(\frac{b}{2a}\right)^2 \text{ på begge sider} \\
\left(x+\frac{b}{2a}\right)^2&=\left(\frac{b}{2a}\right)^2-\frac{c}{a} \quad &\text{Bruker 1. kvadratsetning} \\
\left(x+\frac{b}{2a}\right)^2&=\frac{b^2}{4a^2} \quad &\text{Multipliserer ut potensen på høyre side} \\
\left(x+\frac{b}{2a}\right)^2&=\frac{b^2-4ac}{4a^2} \quad &\text{Setter høyre side på felles brøkstrek} \\
x+\frac{b}{2a}&= \pm \sqrt{\frac{b^2-4ac}{4a^2}} \quad &\text{Tar kvadratrot på begge sider} \\
x &= -\frac{b}{2a} \pm \sqrt{\frac{b^2-4ac}{4a^2}} \quad &\text{Trekker fra } \frac{b}{2a} \text{ på begge sider} \\
x &= -\frac{b}{2a} \pm \frac{\sqrt{b^2-4ac}}{2a} \quad &\text{Regner ut kvadratrot av nevneren} \\
x &= \frac{-b\pm \sqrt{b^2-4ac}}{2a} \quad &\text{Setter på felles brøkstrek}
\end{align}
$
```

Det er ikke nødvendig å huske beviset for abc-formelen, men prøv og forstå alle stegene som er involvert.

Videre skal vi se på en del eksempler på bruk av abc-formelen.

```{admonition} $x^2+20=12x$
:class: eksempel

Vi starter med å trekke fra $12x$ på begge sider og får $x^2-12x+20=0$. Da er likningen på formen $ax^2+bx+c=0$ som vi må ha for å kunne bruke abc-formelen. Vi ser videre at

$$a=1 \quad b=-12 \quad c=20$$

Vi setter inn i abc-formelen

$
\begin{align}
x &=\frac{-(-12)\pm\sqrt{(-12)^2-4\cdot 1\cdot 20}}{2\cdot 1}=\frac{12\pm\sqrt{144-80}}{2} \\
&=\frac{12\pm\sqrt{64}}{2}=\frac{12\pm 8}{2}
\end{align}
$

$
\begin{align}
x=\frac{12+8}{2} \quad &\vee \quad x=\frac{12-8}{2} \\
x=10 \quad &\vee \quad x=2
\end{align}
$

Løsningen blir $L=\{2, 10\}$ tilsvarende som vi fant ved faktorisering i et tidligere eksempel.
```

```{admonition} $2x^2-2x-12=0$
:class: eksempel
Her kan vi bruke $a=2$, $b=-2$ og $c=-12$, men vi kan gjøre det enklere for oss ved å først dividere med $2$ på begge sder av likningen. Da får vi $x^2-x-6=0$ slik at vi kan bruke

$$a=1 \quad b=-1 \quad c=-6$$

Vi setter inn i abc-formelen

$\displaystyle{x=\frac{-(-1)\pm\sqrt{(-1)^2-4\cdot 1\cdot (-6)}}{2\cdot 1}=\frac{1\pm\sqrt{1+24}}{2}=\frac{1\pm\sqrt{25}}{2}=\frac{1\pm 5}{2}}$

$
\begin{align}
x=\frac{1+5}{2} \quad &\vee \quad x=\frac{1-5}{2} \\
x=3 \quad &\vee \quad x=-2
\end{align}
$

Løsningen blir $L=\{-2, 3\}$.
```

```{admonition} $x^2+6x+9=0$
:class: eksempel
Bruker abc-formelen med $a=1$, $b=6$ og $c=9$.

$\displaystyle{x=\frac{-6\pm\sqrt{6^2-4\cdot 1\cdot 9}}{2\cdot 1}=\frac{-6\pm\sqrt{36-36}}{2}=\frac{-6\pm\sqrt{0}}{2}=\frac{-6}{2}=-3}$

Vi ser at vi har bare en løsning $x=-3$ som henger sammen med at uttrykket på venstre siden av likningen er et fullstendig kvadrat $(x+3)^2$.
```

```{admonition} $x^2+4x+5=0$
:class: eksempel
Bruker abc-formelen med $a=1$, $b=4$ og $c=5$.

$\displaystyle{x=\frac{-4\pm\sqrt{4^2-4\cdot 1\cdot 5}}{2\cdot 1}=\frac{-4\pm\sqrt{16-20}}{2}=\frac{-4\pm\sqrt{-4}}{2}}$

Det er ikke mulig å regne ut $\sqrt{-4}$. abc-formelen gir oss derfor ingen løsninger. Utledningen av abc-formelen er helt generell og når abc-formelen ikke gir løsning, så har ikke likningen noen løsning.

Vi skriver $L=\text{Ø}$, der $\text{Ø}$ står for den tomme mengde. Det betyr at likningen ikke har løsning.
```

## Antall løsninger

I eksemplene over har vi sett andregradslikninger med to, en og ingen løsninger. Fra kvadratrotuttrykket i abc-formelen kan vi avgjøre hvor mange løsninger en andregradsliknig har.

```{admonition} Antall løsninger
Uttrykket $b^2-4ac$ som står under kvadraroten i abc-formelen kalles diskriminanten. Diskriminanten bestemmer antall løsninger.

 * $b^2-4ac>0$ gir to forskjellige løsninger
 * $b^2-4ac=0$ gir en løsning (to like løsninger)
 * $b^2-4ac<0$ gir ingen løsning
```
