# Funksjoner i GeoGebra

I mange oppgaver knyttet til funksjoner vil det være praktisk å bruke digitale hjelpemidler for å undersøke funksjonen. Med GeoGebra kan vi tegne grafen og undersøke den visuelt. Det er også mulig å bruke CAS til å undersøke både spesielle og generelle funksjonsuttrykk. Vi starter med å se på de viktigste funksjonene i grafvinduet.

## Graftegning i GeoGebra

 - Skrive inn funksjonsuttrykket
 - Lese av funksjonen langs $x$-aksen
 - Lese av funksjonen langs $y$-aksen
 - Finne eventuelle nullpunkt
 - Finne eventuelle ekstremalpunkt

I tillegg tar vi for oss

 - Finne asymptoter (rasjonale funksjoner)
 - Finne gjennomsnittlig og momentan vekstfart

```````{admonition} Eksempel: Graftegning i GeoGebra
Vi ønsker videre å finne funksjonsverdien for $x=2$ og finne hvilke $x$-verdier som gir $f(x)=5$.

**Lese av x=2:**

Det enkleste er å bare skrive inn $f(2)$ i algebrafeltet. Dersom vi vil vise det grafisk, så skriver vi inn $x=2$ og velger videre "Skjæring mellom to objekt" fra menyen for å finne skjæringspunktet $A=(2, -8)$ mellom grafen til $f(x)$ og linja $x=2$. Det betyr at $f(2)=-8$.

**Lese av f(x)=5:**

Vi skriver inn $y=5$ og bruker "Skjæring mellom to objekt" til å finne skjæringspunktene $B=(-1.82, 5)$, $C=(0.6, 5)$ og $D=(3.22, 5)$ mellom grafen til $f(x)$ og linja $y=5$. Det betyr at $x\in\{-1.82, 0.6, 3.22\}$ gir $f(x)=5$.

`````{admonition} Finne gjennomsnittlig og momentan vekstfart
:class: losning, dropdown

Vi starter med å sette inn punktene på grafen. Den enkleste måten å gjøre det på er ved å skrive $(-2, f(-2))$ og $(0, f(0))$. Så bruker vi linje-verktøyet til å lage ei rett linje gjennom punktene. Til slutt bruker vi verktøye "Stigning" for å få tegnet inn stigningstallet (det kan også leses av likninga for linja i algebrafeltet). Vi får da at gjennomsnittlig vekstfart i $[-2,0]$ er $a=6$.

For å finne momentan vekstfart i $(0, f(0))$ bruker vi det punktet vi allerede har. Videre bruker vi verktøyet "Tangent" for å få tegnet inn tangenten i punktet. Til slutt bruker vi verktøyet "Stigning" for å få tegnet inn stigningstallet som er gitt ved $b=-10$.
```````
